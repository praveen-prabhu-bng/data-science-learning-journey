### Arrays

Let's look at the 2 of the most common structures we see in programming a) The Array & b) The Dictionary.

In Python an Array is referred to as **List**, in JavaScript it's an Array. We will use List & Array interchangeably when referring to this structurem

Definition:
Array (list) -  a linear data structure with a finite number of values . These values can be anything from numbers, texts, other lists, custom objects etc.
In some Languages, arrays enforce only one data type. In python, a list can contain anything (numbers, text, lists, objects ) 

e.g. [ 10,20,30,55,6,67…..]

Arrays are indexed, which means an integer number is assigned to each element of the array, starting from 0,1,2…
We have 2 different ways of storing a list, memory allocation ….

Dynamic array 
- don't need to know the size when creating it 
- it requires re-sizing 

Static array 
- more efficient with memory 
- we have to know the size of the array

### Hash Tables(Dictionaries)

Dictionaries are another very common and incredibly useful data structure we have at our disposal in Python, its name between programming languages varies even more than arrays. They are known as Hash tables, hash maps, Associative arrays or simply maps

Just like we can't have 2 same indexes in Lists, similarly we can't have 2 same Keys in Dictionaries.
We access each values using the keys, similar to accessing each list value using Indexes
