### Introduction to Doubly Linked List 

Single Linked Lists design was perfectly fine for most operations, but it showed its weakness when we wanted to remove the last node, because at that point we needed to know what was the node right before the tail, which could not be accessed directly. So we needed to traverse the entire list starting from the head. This might not always be a huge issue, but for large lists or if we need to do this action, often we could have some performance issues

The solution for this was to modify the node objects themselves, instead of them holding only the value and a pointer to the next node. Why not also add a pointer to the previous node? And this is how the doubly linked list was born

A **Doubly Linked List** is composed of nodes that contain references to both the next node as well as the previous one

The only difference is that each node will have an additional property for the previous node. And that's pretty much all the theory behind it. Let's now go ahead and implement a doubly linked list

For us to transform our previous list into a doubly linked list, we only need to add one new attribute to the node class. So we will just copy and paste the two class definitions, including the constructor functions. And in the node class just add a new property called previous with an initial value of none and everything else inside it will stay the same. As you can see, the base structure of our doubly linked list is almost the same as the singly linked one, and the methods will also be very similar. We only need to adjust them to handle the new property. And that is it. Our setup is complete

```python
class Node: # Creating a Node and initializing it with 2 containers (a.Value  b.Pointing to next/previous node)
    def __init__(self, value):
        self.value = value # This will store the actual value
        self.next = None # this will store the pointer or points to next node it will be linking. For now its None (Node is empty)
        self.previous = None # this will store the pointer or points to previous node it will be linking.
        
class DoublyLinkedList: # Creating a new class to track first(head) and last(tail) element of the List
    def __init__(self): # These are attributes (sometimes called instance variables).They hold values for head, tail & length
        self.head = None # We don't want to give them any starting values.We want the list to be empty when initialized
        self.tail = None # Both the head and the tail will eventually be some node objects we just defined up (in Node Class)
        self._length = 0 # Initial value is set to 0, we will increase and decrease it as we add or remove nodes
# The length is not mandatory, but it will make our lives easier with certain methods.And generally the length of a list is a pretty useful thing to know
# We'll define our append method inside the Doubly linked list class, so that we can call it on an instance of this class    
```

### Append

**Adding new Node**

![](DS_Images/DLL3.png)

**Post Appending New Node**

![](DS_Images/DLL4.png)

```python
def append(self, value): # It takes self (the current LinkedList object) and value (the data you want to add/append).
    new_node = Node(value) # Creates a new Node object with the given value.
    if not self._length: # Checks if the list is empty. "self._length" tracks how many nodes are in the list. If _length is 0, this condition is True
        self.head = self.tail = new_node # If the list is empty,the new node becomes both the head(first node)& the tail(last node).
    else :                               # This initializes the linked list with its first element.
        self.tail.next = new_node # If the list is not empty: links the current last node to the new node.
        new_node.previous = self.tail # link the back pointer to previous node to complete the handshake
        self.tail = new_node # updates the tail pointer so the new node is now the last node.
    self._length += 1 # Increases the length counter by 1 since a new node was added.
    return self # Returns the linked list object itself (self). This allows method chaining* (e.g., mylist.append(10).append(20)
```

We keep everything same as in Single Linked List and add a new line (new_node.previous) we make the tail point to the new node, which is correct, and then we make the new node a new tail. So we still want both of these lines in our new version of append. But we're missing one thing. And that is to make the new node point to the old tail. So before we update the tail, we want a new line that sets the previous property of the new node equal to the tail. And now we have that double connection we just talked about

### Prepend

**Before Append**

![](DS_Images/DLL5.png)

**After Append**

![](DS_Images/DLL6.png)

```python
def prepend(self, value): # Defines a method named inside the class. self refers to the current linked list object.
    new_node = Node(value)
    if not self._length: # Checks if the list is empty (_length == 0).
        self.head = self.tail = new_node#If the list is empty,both head & tail point to the new node since its simultaneously the first & last element.
    else : # Runs if the list is not empty.We need to insert the new node before the current head.
        new_node.next = self.head # Links the new node to the existing list.The new node’s next pointer now points to the old head.
        self.head.previous = new_node # The current head (8) now points back to the new node (5) using its previous pointer.
        self.head = new_node # Updates the list’s head to the new node. Now the new node becomes the first element in the list.
    self._length += 1 #  Increases the length counter by 1.
    return self
```

**Keynotes:**

The function call determines insertion:

	prepend(value) → Adds before the current head.
	append(value) → Adds after the current tail.
	
How does the code know where to insert the node?
It depends on whether you're:

	Prepending (adding at the start) → The new node replaces self.head.
	Appending (adding at the end) → The new node replaces self.tail.
Each function handles pointer updates differently, ensuring correct placement.

Prepend moves backward (right-to-left connection) since we update head.
Append moves forward (left-to-right connection) since we update tail.
The code doesn't "choose" a direction—it follows logical pointer assignments based on where the new node is being added!

### Pop Left

`pop_left` is an operation that **removes the first node (the head) from the list**.  
- The head pointer is updated to the next node.  
- The new head’s `.previous` pointer is set to `None`.  
- The old head node is discarded.  

In short: **`pop_left` deletes the leftmost element and shifts the head forward.**


**Before Pop Left**

![](DS_Images/DLL7.png)

**After Pop Left**

![](DS_Images/DLL8.png)

If singly linked version, we had to update the head property first, and only then could we remove the node. So the order would be like this. First, save the current head into a new variable. Then set the head property to be the next node that comes after the current head. After that, disconnect the former head from the list by setting both its next property and the new head's previous property to none. And lastly, decrement the length of the list. You also need to be mindful of the edge case, like the list being empty

```python
# This method removes and returns the value of the node at the head (leftmost element)
def pop_left(self): # No  parameter is needed because we’re not adding anything, only removing.
    if not self._length:# If the list is empty, we cannot pop anything.Raises an exception to signal the error.
        raise Exception("List is Empty")  # Best practice: always guard against operations on empty structures.
    former_head = self.head # Stores the current head node in a temp variable, so we can return its value later.This is the node we’re about to remove
    if self._length == 1: # Special case: if the list has only one node. Removing it means both head and tail become None. The list is now empty.
        self.head = self.tail = None 
    else :
        self.head = former_head.next # Moves the head pointer to the next node.Effectively removes the old head from the chain.
# If there was only one node, former_head.next is None, so head becomes None
        former_head.next = None # Breaks the link from the removed node to the rest of the list.
        self.head.previous = None # clear backward link of the new head (since it’s now the first node).
    self._length -= 1 # Decrement the length counter by 1. Keeps the _length attribute consistent with the actual number of nodes.
    return former_head.value # Returns the value stored in the removed node.This is the “pop” result, just like list.pop(0) in Python’s built-in list.
```

But remember when we did the singly linked list version of this method, we had to consider the case where the list had only one element because the tail was still pointing to the removed node. It's the same story for this version. For a list with one element, we don't update the tail anywhere. 

Moreover, at this line, we update the head to be its next element, which will be none if the list is only one element long. And when this line runs, we're trying to access the previous property on none, which will break everything. So we definitely need to handle this case. 

There's more than one way of doing it, but the one that I can think of that's quite simple would be to do an if statement here that checks if the length is equal to one, in which case we set both the head and the tail to be none, and then do an else clause where we add all these other lines. This is slightly different from how we did it in the singly linked list, where we checked the length at the end and made the tail equal to none. Both solutions are valid though it's just a matter of preference. What to use. All that's left for this implementation is to decrement the length by one, and then return the value that was inside the former head. And we're done

### Pop Right

In this method we remove the last element from our list. And just like all the other times we removed a node, we do that by severing the connection between it and the rest of the list. 

In the case of the tail, that connection was done through the second to last node. If you recall when we did this for our singly linked list, in order to reach the second to last node, we had to traverse the entire list starting from the head. 

The question now is, do we want to do the same thing with the doubly linked list? Is there no other way of reaching the second to last node? Luckily this time there is, thanks to the previous pointer that all the nodes have. We can now visit the tail directly, then access the node before it using its previous property. And that saves us from all the looping we had to do before

**Before Pop Right**

![](DS_Images/DLL9.png)

**After Pop Right**

![](DS_Images/DLL10.png)


**Procedure:**

Save the tail in a variable. Just like we did when we removed the head. Reassign the tail of the list to be the second to last node by using the former tails. Previous property. After that, remove the old tail by cutting the two links it has to the rest of the list. Lastly, subtract one from the length since we're now minus one node. Also, don't forget to handle the other cases like an empty list or one with a single element. And that is the entire algorithm

```python
def pop_right(self): # pop_right mirrors Python’s list.pop() but specifically from the right end
    if not self._length: # self._length tracks the number of nodes. If _length == 0, the list is empty → popping is invalid.
        raise Exception("The List is Empty") # Raising an Exception prevents undefined behavior
    former_tail = self.tail # Before modifying the list,save the value of the last node.This way we can return it later,even after removing the node
    if self._length == 1:# If there’s only one node: Both head & tail point to the same node.Removing it means the list becomes empty → set both to None.
        self.head = self.tail = None # This is a base case that avoids traversing unnecessarily.
    else : # General Case (More than One Node)
        self.tail = former_tail.previous # Update Tail Reference - Now former_tail.previous(5) becomes the new tail. 
        former_tail.previous = None # Break Link to Old Tail - Disconnect the old tail node.Ensures the list ends cleanly at the new tail.
        self.tail.next = None # Break Link to Old Tail - Disconnect the old tail node.Ensures the list ends cleanly at the new tail.
    self._length -= 1 # Decrement the length counter to reflect removal, Maintaining _length is crucial for O(1) size checks
    return former_tail.value # Finally, return the value of the removed node. This matches Python’s list.pop() behavior.      
```

### Remove

**Summary of Doubly Linked List (DLL) Remove:**

• 	Purpose: Deletes a node containing a specific value (or at a specific position, depending on implementation).

• 	Steps involved:
1. 	Check empty list: If the list has no nodes, raise an error or return nothing.
2. 	Find target node: Traverse from head (or tail) until the node to remove is found.
3. 	Update links:
   - If removing the head, move `head` to `head.next`and clear its `previous`.
   - If removing the tail, move `tail` to `tail.previous` and clear its `next`.
   - If removing a middle node, link its `previous` to its `next` and vice versa, effectively bypassing the node.
4. Detach node: Clear its `next` and `previous` references so it’s fully disconnected.
5. Update length: Decrement `_length`  to keep size accurate. .
6. 	Return value (optional): Often returns the removed node’s value for usability.

```python
def remove(self, value):# Purpose: find the node containing "Value", remove it from the list, and return its value.
    if not self._length:# If _length==0,the list has no nodes.Trying to remove from an empty list is invalid → raise an exception.
        raise Exception("List is Empty")# Best practice: use "ValueError("List is empty")" or "IndexError" for consistency with Python’s built-ins.
    if self.head.value == value: # If the value is in the head node, remove it using "pop_left"(your existing method).
        return self.pop_left()# This avoids duplicating logic for head removal.Design choice: reuse existing methods for consistency and maintainability.
    previous_node = self.head # previous_node tracks the node before current_node
    current_node = self.head.next # Start traversal from the second node (head.next). This setup allows you to "unlink" the target node when found
# Traverse Until Value Found. Loop through nodes until a) You reach the end (current_node is None), OR b) You find the node with the matching value
    while current_node is not None and current_node.value != value: 
        previous_node = current_node # Each iteration: - Move previous_node forward to current_node and 
        current_node = current_node.next # - Move current_node forward to the next node. This is the classic search loop in linked lists
# If while loop exits,its for one of the 2 reason.if its Reason1:Current node is none & we have reached the end & havent found the value, execute below code       
    if current_node is None: # Value Not Found : - If traversal ends (current_node == None), the value doesn’t exist.
        raise ValueError("Item not in the list") # - Raise ValueError to signal the item wasn’t found.
# If its reason 2 : We have a match, so next we check if its the last node (current_node.next is None), we can simply use pop_left to remove tail node.
    if current_node.next is None: # Check If Tail Node : - If current_node.next is None, then current_node is the tail node.
        return self.pop_right()#Removes the last node using a predefined function.We found the value and it was not the tail,we will now proceed to surgically remove it from the list.
# If its reason 2 but not tail,then the value is somewhere in the middle and we need to surgically extract the value, sever connections and join others
    # Next 2 line of codes help connect previous node(5) pointers with current.next node(14) pointers together, completely skipping current node      
    current_node.next.previous = previous_node # Updates the next node’s(14) previous pointer(<-) which is (<-14) to bypass the removed node.
# Instead of removing the chosen node. so this line of code makes Node 14 point to Node 5
# just cutting the pointers.Which means next node points back to previous Node,omitting the current node(14 points to 5) as shown in 2nd pic above
    previous_node.next = current_node.next # Adjusts previous_node to directly skip current_node and point to node after current node.(5 points to 14)
    current_node.previous = None # Clears the removed node’s previous pointer to detach it
    current_node.next = None # Clears the removed node’s previous pointer to detach it
    self._length -= 1 # Decrement the list size counter, Keeps _length consistent with actual number of nodes.
    return current_node.value # Returns the value of the removed node as confirmation.
```

About the complexities. Until now, all of our doubly linked list methods had the best possible performance O(1), both in terms of time and memory utilization. From a memory utilization perspective, this function doesn't care about the growth of the list. It will always only add two new variables to memory, and it will just reassign them inside the loop. The space complexity is therefore an ideal O(1)

### Table of Complexity Comparison - Array vs Singly Linked List vs Doubly Linked List

![](DS_Images/DLL14.png)

We noticed by looking at this table that the doubly linked list excels at one thing, it has the best possible performance on adding and removing from both ends. It's better than the singly linked list and much better than the regular array. So you might ask yourself, why would you ever want to use the singly linked list if the doubly is overall better? 

Well, like always, there are some tradeoffs. Doubly linked lists perform better in more cases, but they also require more memory. And that's because of the nodes it's built with. Each node has to carry one additional pointer. we know how they're stored in memory, more pointers means more space consumed. This can become noticeable for very large linked lists, where a significant amount of memory is allocated for the pointers. On the other hand, in today's day and age, devices generally have enough memory. So again, it's something you need to decide based on the problem you're trying to solve and whether or not memory is an issue for you

We can say that linked lists overcome some limitations of the regular array, and they do that by storing the data in node objects. This allows fast insertion and deletion from the start of the list, and in the case of the doubly linked list from the end as well. In a normal array, those operations would mean shifting all the existing elements to adjust for the new state of the list, and that would impact the entire array. But arrays also have some advantages over the linked lists. Because they use indices, we can immediately access the element at any position just by providing its index. In a linked list we can't do that. We have to navigate through the entire list to reach the desired nodem

### Real-Life Applications of Linked Lists


#### Doubly Linked List (DLL)

- **Web Browsers (Back/Forward Navigation)**
  - Each page visited is a node; allows moving backward or forward easily.


- **Music & Video Players**
  - Playlists allow moving to the next or previous track/video seamlessly.

    
- **Undo/Redo Functionality**
  - Text editors and design tools use DLLs to move back and forth between states.

    
- **Photo Galleries**
  - Navigate images forward/backward in a collection.

    
- **Cache Implementation (LRU Cache)**
  - DLL + HashMap used in Least Recently Used (LRU) cache design.
  - Examples: Browser cache, memory caching in systems.

    
- **Gaming (Deck of Cards, Board Games)**
  - Representing cards or moves where bidirectional traversal is needed.

#### Key Takeaways

- **SLL** → lightweight, forward-only traversal, good for queues, stacks, memory blocks.
- **DLL** → heavier (extra pointer per node), but supports bidirectional traversal and faster deletions.
- Choice depends on whether you need **simplicity & memory efficiency** (SLL) or **flexibility & bidirectional navigation** (DLL).
