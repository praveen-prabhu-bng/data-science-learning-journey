### Introduction to Singly Linked Lists

Although this is one of the most popular data structures that comes out of the box in most programming languages, when we got into the details of its technical implementation, we saw that it had a few limitations.
- The dynamic arrays require memory reallocation whenever we increase them past a specific limit, and that can be computationally expensive.
- In the case of static arrays, we have to know their maximum size when defining them, which can be a problem since we don't always know how many elements we'll end up having
- Also because of the way it's stored in memory  Adding and removing an item from the beginning or from inside an array will cause all the other items  to be shifted, which again will require extra computing time.


This is where the linked list data structure comes into play. Its design naturally overcomes these limitations, and the way it achieves this is by changing how elements are stored in memory.

- A linked list representation of the same collection of numbers will look something like this. Each element will have its own container. And each element will have a connection to the next one after it
- It's important to note that each element will have only one link to the next element.
- So when representing this in code, we'll have to create a container with two pieces, one for the information we want to store (orange) and one for the link which will visually represent as this arrow (white)
- And using this arrow, we can connect the current item to the next one and then the next one, and so on
- The last element will also be one of these containers that will contain an arrow. But what will it point to? Logically, if it's the last element, it means that it doesn't point to any other value. It points to, well, nothing. And in Python we represent the concept of nothing with the none value.

**Why Classes Are Needed for Linked Lists ?**
1. Encapsulation of Data + Behavior
- A linked list isn’t just data; it’s also the operations you perform on it (append, delete, search, traverse).
- Using a class lets you bundle the nodes (data) and the methods (behavior) together in one structure.

2. Node Representation
- Each element in a linked list is a Node (with value and next pointer).
- A Node class makes it easy to define this structure once and reuse it:

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

3. LinkedList Management
- The linked list itself needs to track:
- head (first node)
- tail (last node)
- _length (number of nodes)

 
- A LinkedList class keeps these attributes organized and provides methods like:
- append(value)
- insert(index, value)
- delete(value)
- __len__() for length

4. Reusability & Modularity
- With classes, you can create multiple linked lists independently:

```python
list1 = LinkedList()
list2 = LinkedList()
# Each list manages its own nodes without interfering with others.
```

5. Cleaner Code & Abstraction
- Without classes, you’d juggle raw dictionaries or tuples for nodes and manually handle pointers — messy and error-prone.
- Classes abstract away the details, so you can just call mylist.append(10) instead of writing pointer logic every time.

```python
# Example Structure
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._length = 0

    def append(self, value):
        new_node = Node(value)
        if not self._length:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self._length += 1
        return self
```

🌟 Key Takeaway

We use classes for linked lists because they:
- Organize nodes and operations together
- Provide abstraction (hide pointer details)
- Enable reusability and modularity
- Make the code cleaner, safer, and easier to extend

This separation follows the principle of single responsibility:
- A Node only knows about itself and its neighbor.
- The SinglyLinkedList knows about the whole chain.

**Steps to append a node**
1. Wrap the value in a node object.
2. Make the current tail point to it.
3. And then update the tail property.

**Let's now see how we can put that in code.**

- The first thing we will do is to create a new node instance using that value, which I will call new node.
- We then said that we want the current tail to point to this new node, and we do that by assigning its next property to the new node.
- And then because our new node is now the last element of the list, we need to update the tail to be equal to this new node.

But there is a problem with our code. Remember we said that when we create a new instance of our linked list, the head and tail will be none by default Now, if we have an empty list and we were to call append on it, this line over here will fail because we're trying to access the next attribute on the tail, which will be none. So we need to write some logic to handle this case. If you think about it, 

when adding an element to an empty list, that single element would be both the start and the end of the list. So for our case, we want to say that the **new node should be both the head and the tail**. And in Python we can assign a value to two variables at the same time using one line just like this. 
And if the list is not empty, we want to execute these two lines we already have. 

This syntax over here is how we would check if our list is empty. We could just check if the head or tail and none. That would be another correct option. 

But what we're saying here is that if the length attribute evaluates to false, and it will evaluate  to false only when it's equal to zero. Execute this next line of code. Recall that the if clause only runs if this condition is true, and in our case not false would evaluate to true. After that we want to increment the length by one.

```python
def append(self, value): # It takes self (the current LinkedList object) and value (the data you want to add/append).
    new_node = Node(value) # Creates a new Node object with the given value.
    if not self._length: # Checks if the list is empty. "self._length" tracks how many nodes are in the list. If _length is 0, this condition is True
        self.head = self.tail = new_node # If the list is empty,the new node becomes both the head(first node)& the tail(last node).
    else :                               # This initializes the linked list with its first element.
        self.tail.next = new_node # If the list is not empty: links the current last node to the new node.
        self.tail = new_node # updates the tail pointer so the new node is now the last node.
    self._length += 1 # Increases the length counter by 1 since a new node was added.
    return self # Returns the linked list object itself (self). This allows method chaining* (e.g., mylist.append(10).append(20)
```

*Method chaining means calling multiple methods on the same object in a single line, one after another. It works when each method returns the object itself (self) instead of None.

```python
class Node: # Creating a Node and initializing it with 2 containers (a.Value  b.Pointing to next node/ next pointer)
    def __init__(self, value):
        self.value = value # This will store the actual value
        self.next = None # this will store the pointer or points to next node it will be linking. For now its None (Node is empty)

class SinglyLinkedList: # Creating a new class to track pfirst(head) and last(tail) element of the List
    def __init__(self): # These are attributes (sometimes called instance variables).They hold values for head, tail & length
        self.head = None # We don't want to give them any starting values.We want the list to be empty when initialized
        self.tail = None # Both the head and the tail will eventually be some node objects we just defined up (in Node Class)
        self._length = 0 # Initial value is set to 0, we will increase and decrease it as we add or remove nodes
# The length is not mandatory, but it will make our lives easier with certain methods.And generally the length of a list is a pretty useful thing to know
# We'll define our append method inside the singly linked list class, so that we can call it on an instance of this class    
    def append(self, value): # This is a method. It’s a behavior: it changes the state of the list by adding a new node
        new_node = Node(value) # Input value is assigned as "new_node" so we continue with next tasks like, updating tail, assigning tail ...etc
        if not self._length: # If the list is empty,the new node becomes both the head(first node)& the tail(last node).
            self.head = self.tail = new_node 
        else :               # This initializes the linked list with its first element.                  
            self.tail.next = new_node  # If the list is not empty: then the pointer of tail(tail.next) is pointed towards new_node+
            self.tail = new_node # And the tail is also updated as "new_node"
        self._length += 1 # regardless of first node/entry or not length is updated for every entry(outside the loop)
        return self # Returns the linked list object itself (self). This allows method chaining* (e.g., mylist.append(10).append(20)
```

### Prepend

Prepend works similarly to append in the sense that we again need to create a node object to store the  value we want to add. So once again, our function will take the value we give it as an argument and create a new node object. Then assign its next attribute to the current head of the list.  And lastly, make the new node the head of the list. And that's it.  These are all the steps that go into it.


**BEFORE**

**AFTER**

![](DS_Images/SLL6.png)

```python
def prepend(self, value): # Defines a method named inside the class. self refers to the current linked list object.
    new_node = Node(value)
    if not self._length: # Checks if the list is empty (_length == 0).
        self.head = self.tail = new_node#If the list is empty,both head & tail point to the new node since its simultaneously the first & last element.
    else : # Runs if the list is not empty.We need to insert the new node before the current head.
        new_node.next = self.head # Links the new node to the existing list.The new node’s next pointer now points to the old head.
        self.head = new_node # Updates the list’s head to the new node. Now the new node becomes the first element in the list.
    self._length += 1 #  Increases the length counter by 1.
    return self
```

### Pop Left

- Pop Left = remove the first element (the head) of the list and return its value.
- It’s called “left” because if you visualize the list as a sequence (like [head → ... → tail]), the head is on the left side.
- After popping, the head pointer moves to the next node.

**BEFORE POP**

![](DS_Images/SLL7.png)

**AFTER POP**

![](DS_Images/SLL8.png)

The trick here is to reassign the head value before we remove the head.  So first we need to save the current head in a variable.  And then we can just move the head attribute to the next element that comes after five. After that, make the old head point to nothing and that would remove it. So once again the steps are. 

- Save the head in a new variable.
- Move the head of the list to the next node.
- Then make the old head we have stored in the variable point to none.

```python
# This method removes the first node(head) of the list and returns its value
def pop_left(self): # No  parameter is needed because we’re not adding anything, only removing.
    if not self._length:
        raise Exception("List is Empty") # If the list is empty, we cannot pop anything.Raises an exception to signal the error.
    former_head = self.head # Stores the current head node in a temp variable, so we can return its value later.This is the node we’re about to remove
    self.head = former_head.next # Moves the head pointer to the next node.Effectively removes the old head from the chain.
# If there was only one node, former_head.next is None, so head becomes None
    former_head.next = None # Breaks the link from the removed node to the rest of the list.
    self._length -= 1
    if not self._length: # Checks again if the list is now empty (after removal).This happens if we just removed the only node
        self.tail = None #If the list is empty, set tail to None as well. Ensures both head and tail are None when the list has no nodes.
    return former_head.value # Returns the value stored in the removed node.This is the “pop” result, just like list.pop(0) in Python’s built-in list.
```

**Why Add the Exception Error?**

If we have an empty list and we call this pop left. We want some kind of response back that tells us there is no element in the list. And now we get a more correct response back. 

We now know the list was empty and there was no element to remove. But I personally like them, and I think it creates a good habit of thinking about all the possible scenarios you could encounter.

**Why append / prepend return self**
- These methods add a new node to the list.
- Returning self allows method chaining:

mylist.append(10).append(20).prepend(5)
    
- You don’t care about the new node itself — you care about the updated list.
- So returning the list object (self) makes sense.

**Why pop_left returns former_head.value**
- pop_left removes a node.
- The important result is the value that was removed, not the updated list.

- Example:

print(val)  # shows the removed value

- If it returned self, you’d lose the information about which element was popped.
- Returning the node’s value is consistent with Python’s built-in list.pop(), which also returns the removed element.

**Analogy**

Think of a vending machine:
- Append/Prepend → you’re loading snacks into the machine. You want to keep interacting with the machine itself (so return self).
- Pop Left → you’re taking a snack out. You don’t care about the machine object at that moment — you care about the snack (so return the value).

✅ So:
- Mutator methods (append/prepend) → return self for chaining.
- Removal methods (pop_left) → return the removed value, because that’s the useful result.

### Pop Right

If pop left was the name of the function that removed the first element. Pop right will remove the last one.

**BEFORE POP RIGHT**

So if we want to remove the tail, we must make 11 point to none instead of to the eight node. That sounds pretty simple, right? But let me ask you this. How do we access the second to last node?

We know that we have access to the last node using the tail attribute, but the tail doesn't have access to its preceding element. In fact, none of our elements do. Recall that a node object only carries a pointer to the next item.   So if we take the node eight as an example, all it knows is that it carries a value and its next attribute is none  It doesn't know who is pointing to it. Who is the preceding node? So how do we access the second to last node then?

The steps for this method would be like this. Loop through all the nodes starting from the head of the list, checking if each node points to the tail, and when you find the one that does have it, point to none instead.  And then make that node the new tail.  But also be mindful of the special cases and the extra bits of work, like decreasing the length and what value you should return.

```python
def pop_right(self): # pop_right mirrors Python’s list.pop() but specifically from the right end
    if not self._length: # self._length tracks the number of nodes. If _length == 0, the list is empty → popping is invalid.
        raise Exception("List is empty") # Raising an Exception prevents undefined behavior
    tail_value = self.tail.value # Before modifying the list,save the value of the last node.This way we can return it later,even after removing the node
    if self._length == 1:# If there’s only one node: Both head & tail point to the same node.Removing it means the list becomes empty → set both to None.
        self.head = self.tail = None # This is a base case that avoids traversing unnecessarily.
    else : # General Case (More than One Node)
        temp_node = self.head # Start from the head to find the second-to-last node. 
        while temp_node.next is not self.tail: # Traverse Until Node Before Tail.Walk through the list until temp_node.next points to the current tail.
            temp_node = temp_node.next # This loop stops at the node just before the tail.
        self.tail = temp_node # Update Tail Reference - Now temp_node becomes the new tail.
        self.tail.next = None # Break Link to Old Tail - Disconnect the old tail node.Ensures the list ends cleanly at the new tail.
    self._length -= 1 # Decrement the length counter to reflect removal, Maintaining _length is crucial for O(1) size checks
    return tail_value # Finally, return the value of the removed node. This matches Python’s list.pop() behavior.
```

So let's have a look at our list. And now, since there is no other node pointing to it, 11 will be removed from the list. So that's what the end result would look like

Let's assume we want to remove the number 11 from the list. When we provide the value 11 as a parameter, Python will save that in a variable and then run the code as usual from top to bottom, checking the length (if the list is empty), then checking the first node (if match then run pop_left). And since none of these conditions are met, we now create the previous and current node variables. We then enter the while loop. 

In the first iteration, current node is 11. It's not the one we're looking for. So we bring the previous node forward to be what the current node is, and then we reassign the current node to be its next node. And then we ask again these two conditions at while loop.And since both are true we keep on moving both variables forward and we verify with each iteration whether we reached the end of the list or we found the node. 

At this point, current node ended up being 11, which is the one we're looking for, so the loop stops ,after we break out. We check again. Did we break because we reached the end? In this case, no. So we moved to the next condition,  is our current node The tail ? is its next attribute None ?. No, it's the 8 node. So then we proceed to remove the element we found. We point the previous node to the current node's next node. In our example point 14 to 8. And then make the current node point to none instead of 8. And finally the last two lines run after that, decreasing the length and returning the value we removed. And that's it

One last thing I want to mention about our function because of how we built the while loop. Our function will only remove the first node it finds with the target value. So if we were to have two nodes that contain the number 14, for example, we don't remove both of them, only the first one, and then the execution stops. This is, however, not something unusual in most programming languages. Methods that remove elements based on a specified value stop when they find the first occurrence, and our version is therefore compliant with this general rule.

```python
def remove(self, value):# Purpose: find the node containing "Value", remove it from the list, and return its value.
    if not self._length:# If _length==0,the list has no nodes.Trying to remove from an empty list is invalid → raise an exception.
        raise Exception("List is Empty")# Best practice: use "ValueError("List is empty")" or "IndexError" for consistency with Python’s built-ins.
    if self.head.value == value: # If the value is in the head node, remove it using "pop_left"(your existing method).
        return self.pop_left()# This avoids duplicating logic for head removal.Design choice: reuse existing methods for consistency and maintainability.
    previous_node = self.head # previous_node tracks the node before current_node
    current_node = self.head.next # Start traversal from the second node (head.next). This setup allows you to "unlink" the target node when found
# Traverse Until Value Found. Loop through nodes until a) You reach the end (current_node is None), OR b) You find the node with the matching value
    while current_node is not None and current_node.value != value: 
        previous_node = current_node # Each iteration: - Move previous_node forward to current_node and 
        current_node = current_node.next # - Move current_node forward to the next node. This is the classic search loop in linked lists
    if current_node is None: # Value Not Found : - If traversal ends (current_node == None), the value doesn’t exist.
        raise ValueError("Item not in the list") # - Raise ValueError to signal the item wasn’t found.
    if current_node.next is None: # Check If Tail Node : - If current_node.next is None, then current_node is the tail node.
        self.tail = previous_node
    previous_node.next = current_node.next # Unlink the Node : - Skip over the current_node by linking previous_node.next directly to current_node.next
    current_node.next = None # Disconnect Removed Node : Break the link from the removed node to the rest of the list
    self._length -= 1 # Decrement the list size counter, Keeps _length consistent with actual number of nodes.
    return current_node.value
```

### Reverse

**Reverse** means flipping the direction of the list so the last node becomes the first, and the first node becomes the last.
- Every node’s .next pointer is changed to point backward instead of forward.
- The head becomes the old tail, and the tail becomes the old head.
- It’s done by walking through the list and carefully reassigning links one by one.

👉 In one line: Reverse in SLL is the process of turning the list around so traversal goes in the opposite direction.

**BEFORE REVERSE**

![](DS_Images/SLL13.png)

**AFTER REVERSE**

![](DS_Images/SLL14.png)

Imagine you have a chain of paper clips hooked together in a line, where each clip points to the next one. This algorithm flips the entire chain around so the last clip becomes first, and the first becomes last!

```python
def reverse(self): # - Purpose: reverse the order of nodes in the list so traversal goes in the opposite direction.
    if self._length < 2 : # If the list has fewer than 2 nodes (empty or single node), reversing doesn’t change anything.
        return self # Simply return the list itself. Efficiency note: avoids unnecessary looping.
# Set Up Three Helper Hands-
    left_node = None # 1.Left hand(left_node):Starts holding nothing (None) (L=None)
    middle_node = self.head#2.Middle hand (middle_node):Grabs the very first clip(the head) (M=08)
    while middle_node is not None :
        right_node = middle_node.next # 3.Right hand(right_node) grabs the next clip in the chain so you don't lose track of where to go next.
# (R=11),(R=11 to 05),(R=05 to 09),(R=09 to 14),(R=14 to None), R = None, breaks out of while loop
# This is where the actual reversal is happening, we are changing the direction of the pointer (from left to right "->" ) to right to left "<-"
        middle_node.next = left_node#Make the middle clip point backwards to the clip in your left hand(instead of forward).
# (M=None<- 08),(M=08<-11),(M=11<-05),(M=05<-09),(M=09<-14)
        left_node = middle_node # Your left hand moves up to grab what the middle hand was holding 
# (L=None to 08),(L=8 to 11),(L=11 to 05),(L=05 to 09),(L=09 to 14)
        middle_node = right_node # Your middle hand moves to what the right hand was holding 
# (M=8 to 11),(M=11 to 05),(M=05 to 09),(M=09 to 14),(M=14 to None)
# After the loop finishes, the chain is reversed, but the list still thinks the old first clip is the head. 
# So we swap the labels: what was the tail is now the head, and vice versa!
    self.head, self.tail = self.tail, self.head # self.head (08), self.tail (14) = self.tail (08), self.head (14)
    return self
```

### Final Algorithm 

```python
class Node: # Creating a Node and initializing it with 2 containers (a.Value  b.Pointing to next node/ next pointer)
    def __init__(self, value):
        self.value = value # This will store the actual value
        self.next = None # this will store the pointer or points to next node it will be linking. For now its None (Node is empty)

class SinglyLinkedList: # Creating a new class to track pfirst(head) and last(tail) element of the List
    def __init__(self): # These are attributes (sometimes called instance variables).They hold values for head, tail & length
        self.head = None # We don't want to give them any starting values.We want the list to be empty when initialized
        self.tail = None # Both the head and the tail will eventually be some node objects we just defined up (in Node Class)
        self._length = 0 # Initial value is set to 0, we will increase and decrease it as we add or remove nodes
# The length is not mandatory, but it will make our lives easier with certain methods.And generally the length of a list is a pretty useful thing to know
# We'll define our append method inside the singly linked list class, so that we can call it on an instance of this class    

    
    def append(self, value): # This is a method. It’s a behavior: it changes the state of the list by adding a new node
        new_node = Node(value) # Input value is assigned as "new_node" so we continue with next tasks like, updating tail, assigning tail ...etc
        if not self._length: # If the list is empty,the new node becomes both the head(first node)& the tail(last node).
            self.head = self.tail = new_node 
        else :               # This initializes the linked list with its first element.                  
            self.tail.next = new_node  # If the list is not empty: then the pointer of tail(tail.next) is pointed towards new_node+
            self.tail = new_node # And the tail is also updated as "new_node"
        self._length += 1 # regardless of first node/entry or not length is updated for every entry(outside the loop)
        return self # Returns the linked list object itself (self). This allows method chaining* (e.g., mylist.append(10).append(20)

    
    
    def prepend(self, value): # Defines a method named inside the class. self refers to the current linked list object.
        new_node = Node(value)
        if not self._length: # Checks if the list is empty (_length == 0).
            self.head = self.tail = new_node#If the list is empty,both head & tail point to the new node since its simultaneously the first & last element.
        else : # Runs if the list is not empty.We need to insert the new node before the current head.
            new_node.next = self.head # Links the new node to the existing list.The new node’s next pointer now points to the old head.
            self.head = new_node # Updates the list’s head to the new node. Now the new node becomes the first element in the list.
        self._length += 1 # - Increases the length counter by 1.Keeps track of how many nodes are in the list.
        return self

    
    
    def pop_left(self): # No  parameter is needed because we’re not adding anything, only removing.
        if not self._length:
            raise Exception("List is Empty") # If the list is empty, we cannot pop anything.Raises an exception to signal the error.
        former_head = self.head # Stores the current head node in a temp variable, so we can return its value later.This is the node we’re about to remove
        self.head = former_head.next # Moves the head pointer to the next node.Effectively removes the old head from the chain.
    # If there was only one node, former_head.next is None, so head becomes None
        former_head.next = None # Breaks the link from the removed node to the rest of the list.
        self._length -= 1
        if not self._length: # Checks again if the list is now empty (after removal).This happens if we just removed the only node
            self.tail = None #If the list is empty, set tail to None as well. Ensures both head and tail are None when the list has no nodes.
        return former_head.value # Returns the value stored in the removed node.This is the “pop” result, just like list.pop(0) in Python’s built-in list.


    
    def pop_right(self): # pop_right mirrors Python’s list.pop() but specifically from the right end
        if not self._length: # self._length tracks the number of nodes. If _length == 0, the list is empty → popping is invalid.
            raise Exception("List is empty") # Raising an Exception prevents undefined behavior
        tail_value = self.tail.value # Before modifying the list,save the value of the last node.This way we can return it later,even after removing the node
        if self._length == 1:# If there’s only one node: Both head & tail point to the same node.Removing it means the list becomes empty → set both to None.
            self.head = self.tail = None # This is a base case that avoids traversing unnecessarily.
        else : # General Case (More than One Node)
            temp_node = self.head # Start from the head to find the second-to-last node. 
            while temp_node.next is not self.tail: # Traverse Until Node Before Tail.Walk through the list until temp_node.next points to the current tail.
                temp_node = temp_node.next # This loop stops at the node just before the tail.
            self.tail = temp_node # Update Tail Reference - Now temp_node becomes the new tail.
            self.tail.next = None # Break Link to Old Tail - Disconnect the old tail node.Ensures the list ends cleanly at the new tail.
        self._length -= 1 # Decrement the length counter to reflect removal, Maintaining _length is crucial for O(1) size checks
        return tail_value # Finally, return the value of the removed node. This matches Python’s list.pop() behavior.

    

    def remove(self, value):# Purpose: find the node containing "Value", remove it from the list, and return its value.
        if not self._length:# If _length==0,the list has no nodes.Trying to remove from an empty list is invalid → raise an exception.
            raise Exception("List is Empty")# Best practice: use "ValueError("List is empty")" or "IndexError" for consistency with Python’s built-ins.
        if self.head.value == value: # If the value is in the head node, remove it using "pop_left"(your existing method).
            return self.pop_left()# This avoids duplicating logic for head removal.Design choice: reuse existing methods for consistency and maintainability.
        previous_node = self.head # previous_node tracks the node before current_node
        current_node = self.head.next # Start traversal from the second node (head.next). This setup allows you to "unlink" the target node when found
    # Traverse Until Value Found. Loop through nodes until a) You reach the end (current_node is None), OR b) You find the node with the matching value
        while current_node is not None and current_node.value != value: 
            previous_node = current_node # Each iteration: - Move previous_node forward to current_node and 
            current_node = current_node.next # - Move current_node forward to the next node. This is the classic search loop in linked lists
        if current_node is None: # Value Not Found : - If traversal ends (current_node == None), the value doesn’t exist.
            raise ValueError("Item not in the list") # - Raise ValueError to signal the item wasn’t found.
        if current_node.next is None: # Check If Tail Node : - If current_node.next is None, then current_node is the tail node.
            self.tail = previous_node
        previous_node.next = current_node.next # Unlink the Node : - Skip over the current_node by linking previous_node.next directly to current_node.next
        current_node.next = None # Disconnect Removed Node : Break the link from the removed node to the rest of the list
        self._length -= 1 # Decrement the list size counter, Keeps _length consistent with actual number of nodes.
        return current_node.value



    def reverse(self): # - Purpose: reverse the order of nodes in the list so traversal goes in the opposite direction.
        if self._length < 2 : # If the list has fewer than 2 nodes (empty or single node), reversing doesn’t change anything.
            return self # Simply return the list itself. Efficiency note: avoids unnecessary looping.
    # Set Up Three Helper Hands-
        left_node = None # 1.Left hand(left_node):Starts holding nothing (None) (L=None)
        middle_node = self.head#2.Middle hand (middle_node):Grabs the very first clip(the head) (M=08)
        while middle_node is not None :
            right_node = middle_node.next # 3.Right hand(right_node) grabs the next clip in the chain so you don't lose track of where to go next.
    # (R=11),(R=11 to 05),(R=05 to 09),(R=09 to 14),(R=14 to None), R = None, breaks out of while loop
    # This is where the actual reversal is happening, we are changing the direction of the pointer (from left to right "->" ) to right to left "<-"
            middle_node.next = left_node#Make the middle clip point backwards to the clip in your left hand(instead of forward).
    # (M=None<- 08),(M=08<-11),(M=11<-05),(M=05<-09),(M=09<-14)
            left_node = middle_node # Your left hand moves up to grab what the middle hand was holding 
    # (L=None to 08),(L=8 to 11),(L=11 to 05),(L=05 to 09),(L=09 to 14)
            middle_node = right_node # Your middle hand moves to what the right hand was holding 
    # (M=8 to 11),(M=11 to 05),(M=05 to 09),(M=09 to 14),(M=14 to None)
    # After the loop finishes, the chain is reversed, but the list still thinks the old first clip is the head. 
    # So we swap the labels: what was the tail is now the head, and vice versa!
        self.head, self.tail = self.tail, self.head # self.head (08), self.tail (14) = self.tail (08), self.head (14)
        return self
```

### Singly Linked List Memory Allocation

Lets review the concept of a linked list.And this time, we will try to picture it in a way that's closer to how it's stored in our computer's memory.

we saw that both the linked list and the nodes are implemented as classes. Therefore, they can be represented as these enclosed containers, each with its own data. The linked list object will be this one that has the length, head and tail properties, and then the nodes objects will be separate containers, each with its own private information and the pointers toother nodes. So the first one will be the head, which will point to a different node, which in turn points to another one, and so on until the last one doesn't point to anything.

To better understand how Linked Lists work .Let's imagine our computer's memory right here to the right. Let's now assume that our links list object is stored in these three memory slots (0x0048,4C,50)

We keep saying that this node points to this node and the head points to that node and so on. But what does that even mean? Well, you can think of a pointer as being a number that represents one of these memory addresses. So when we say that the head of the list is equal to a certain node or points to a certain node, what I actually mean is that in the section where our linked list object is stored, the value of the head attribute will be the memory location of our first node, let's say at number 74.

So when we access the head attribute, we get whatever is stored in location 74, which in our case will be a different object, this time a node object, and it too will have some information stored in it, like the value which will be whatever we want it to be, and the next attribute. And guess what this next attribute will be? You guessed it. It will also be a memory location of another node. And the thing is, it doesn't have to be the next one right after it. Therefore, in the memory block where our list object is stored, the last allocated address will contain the location of this node. 

Hopefully right now you can see clearer what all these operations meant when we needed to traverse the list.By going from one node to another, we just access the memory location indicated by the next property of each node object. ### Table of Complexity Comparison - Array vs Singly Linked List

The first two are adding and removing from the start of the classic array. Both these operations need O(n) time to run, because if you remember, we need to reassign all the following elements to different indices. Adding and removing from the end will run in constant time, with the mention that sometimes adding to the end will require the entire array to be copied to another memory location, and in that case the runtime will be O(n), but since that happens very rarely with large arrays, the generally accepted complexity is O(1). Searching will be O(n) time, since you could potentially run over all the elements of the list until you find the one you're looking for. And such iteration will require linear time & access. This is done instantly in a normal array because it is indexed, and we can get the element at any position just by referencing its index. 

![](DS_Images/SLL17.png)

If we now look at the singly linked list. The first two operations, unlike the classic array, are both simple functions. No re-indexing is required for the linked list. Both run in constant time. Adding to the end is also constant time, because we have a direct pointer to the tail node, and all we need to do is attach it to the last node. Removing from the end, on the other hand, is done in O(n)  time. Since we're forced to go over all the nodes to reach the second to last node and then cut the link to the last node. Searching is also anO(n)  operation, just like the normal array.

 If we want to find an element with a specific value, we don't really have a choice other than checking them one by one. Lastly, unlike the classic array in the linked list, access is done in linear time. If we want to know the value of the 50th node, for example, we don't have direct access to it like we do with indices, so we'll have to traverse one node at a time until we reach it.

### Real-Life Applications of Linked Lists

#### Singly Linked List (SLL)

- **Dynamic Memory Management**
  - Used in operating systems to manage free memory blocks.
  - Efficient allocation/deallocation without requiring contiguous memory.


- **Stacks & Queues**
  - Natural fit for implementing stacks (LIFO) and queues (FIFO).
  - Examples: Job scheduling in OS, printer queues.


- **Polynomial Arithmetic & Large Numbers**
  - Represent polynomials or very large integers where each node stores a coefficient/digit.
  - Useful in scientific computing and cryptography.


- **File Systems & Symbol Tables**
  - Directory structures and symbol tables in compilers often use linked lists.


- **Sparse Matrices**
  - Efficient representation of sparse data (lots of zeros) in scientific applications.
