### Introduction to Stacks

So up until now, when we discuss data structures, we always thought about them in terms of different ways of storing data. Regular arrays are stored with indices and linked lists with nodes. Stacks and queues are different in that regard

**Linear data structure** : 
A data structure is defined as linear when it has its elements arranged in a sequence one after another, all on the same level.

![](DS_Images/Stacks1.png)

This means that a linear data structure has two possible extremities, two ends. Examples of such data structures are arrays, singly linked lists, and doubly linked lists. So all of those which we learned until now can be described as being linear. They're all arranged in a sequential manner

**Stacks** :

A stack can be any one of these linear data structures, as long as it follows one rule, and that is to allow adding and removing elements from only one of its ends

![](DS_Images/Stacks2.png)

Oftentimes, stacks are visualized vertically, with the entry and exit point being the top. The actions of adding and removing elements are called `Push` and `Pop`. 

We push new elements onto the stack. And then we pop them off the stack. You can imagine this action as if you were looking down from above the stack, and you can only see and access the top element. Some people like to compare Stacks with the real life example of a stack of trays in the cafeteria. When we want to add more trays, we put them on top of the existing ones, and when we take a tray, we take the top one. 

But whether you like to imagine a stack vertically or horizontally, the only rule you need to remember is that **Stacks only use one end for accessing elements**. And regardless of whether you use regular arrays, singly linked lists, or doubly linked lists, as long as you follow this rule, you have a stack

### Stacks - Array Implementation

A stack operates like a pile of plates (LIFO - Last-In, First-Out)—the last one placed on top is the first one removed. In programming, this means that the most recently added element is the one accessed first. You use push (to add) and pop (to remove) operations.

As discussed before we need to use either the start of the list as the entry point or the end, but not both
Recall that any operation done at the beginning of an array has an O(n) complexity (refer to snapshot below for Arrays), because we have to re index all the following items. This is a problem we don't have when we add or remove from the end, since both operations are O(1) .So naturally we want our code to perform well. So if you're planning to use a regular array as a stack, use its end for all operations

![](DS_Images/DLL14.png)

**Array-Based Stack Implementation**

In Python, stacks can be implemented using lists since lists provide dynamic resizing and support stack operations efficiently.

- Operations in an Array-Based Stack
   1. Push – Adds an item to the top of the stack.
   2. Pop – Removes and returns the top item.
   3. Peek – Views the top item without removing it.
   4. isEmpty – Checks if the stack is empty.


**Python Implementation Using Lists**

```python
class stack:
    def __init__(self): # - Initializes an empty list to represent the stack.
        self.stack = [] # Best practice: Consider prefixing with  if you want to emphasize privacy (though Python doesn’t enforce it).

    def push(self,item): # Adds an item to the top of the stack.
        self.stack.append(item)

    def pop(self): # Removes and returns the top item
        if not self.is_empty():
            return self.stack.pop()
        else : # - Improvement: Instead of raising a generic Exception, raise a more specific one like IndexError or create a custom StackEmptyError.
            raise Exception ("Stack is empty")

    def peek(self): # Returns the top item without removing it.
        return self.stack[-1] if not self.is_empty() else None  # single-line conditional return whch is equivalent to code below in comments

    def is_empty(self): # Checks if the stack is empty.
        return len(self.stack) == 0


#  if not self.is_empty():
#      return self.stack[-1]
#  else:
#      return None
```

### Stacks - Linked Lists Implementation

Let's try to implement a stack using linked lists. And we have two options to choose from.We can either use singly linked lists or doubly linked lists.Both can be adapted to be viable stacks in a singly linked list.

Adding to the end is an O(1) operation, but removing (using while loop) is an O(n), because if you remember, we had to traverse the entire list.Adding and removing from the head of the list, on the other hand, are both constant time operations.So if we choose a singly linked list, we'll have to use the head of the list as the working end of the stack.

![](DS_Images/DLL14.png)

A doubly linked list has O(1) performance on the operations done on both extremities, but we just learned that they also take up more memory than singly linked lists.So between the two I'll go with singly linked lists because we only use one end of the list.We don't care about having the best possible performance on both ends, and we never use the extra pointer so we can save a bit of memory.

So if our stack is based on a linked list, we need to first define its building blocks, which are the `node` objects.Lets just bring the node class we used in the singly linked list section and use it in our stack.
Let's also move to the list class, which we will simply name `stack`

Inside the init method we previously had the head, tail, and length properties.But for our stack we need to adapt a bit.We don't need to keep track of both ends, we only need the head, since that will be both the entry and the exit point of our stack (since its O(1) in Operation).And since most of the time a stack is pictured vertically, we can call this property `top` to be more suitable for a stack.We can also add a property that reflects the length of our stack.Previously we called it length, but here we will go with `size`, which will start at zero.


Another thing we could do is to impose a limit on our stack. Some stacks have what's called a bounded capacity, meaning they can only grow up to a predefined size. Afterwards, they cannot accept new elements. This is again not applicable for all stacks. It's just a variation on the regular stack. 

#### Analogy on List‑based stack vs Linked‑list stack
- List‑based stack: Like a bookshelf — you can put books on top and remove them, but the shelf itself keeps track of order and size.
- Linked‑list stack: Like a chain of paperclips — each clip (Node) holds the next one. You need to remember where the chain starts (_top) and how many clips are in the chain (_size).


```python
# In a linked list implementation, each element must store a)Its value & b)A reference (pointer) to the next element. Without Node, you wouldn’t have 
# a way to “link” elements together. Think of Node as a chain link: each link holds data and points to the next link.

class Node: # Building Block of the Stack. This class enables the linked structure — each node links to the next, forming the stack.
    def __init__(self, value): # Constructor to initialize the node with a value.
        self.value = value
        self.next = None

class Stack: # Class — Linked List-Based Stack
    def __init__(self):
        self._top = None # Points to the top node of the stack (initially empty).
        self._size = 0  # Tracks the number of elements in the stack.
        self._max_allowed_size = 10 # Optional size limit to prevent overflow.

    def push(self, value): # Method — Add Element to Top. This method maintains O(1) time complexity and preserves stack order.
        if self._max_allowed_size == self._size:
            raise Exception ("Stack Size Exceeded")
        new_element = Node(value) # Creates a new node with the given value.
        new_element.next = self._top # Links the new node to the current top.
        self._top = new_element # Updates the top pointer to the new node.
        self._size += 1 # Increments the size counter.
        return self # Enables method chaining (e.g., stack.push(1).push(2)).

    def pop(self): # Method — Remove and Return Top Element.This method ensures safe removal and maintains stack integrity.
        if not self._size:
            raise Exception("Stack is Empty")
        former_top = self._top # Stores the current top node.
        self._top = former_top.next # Moves the top pointer to the next node.
        former_top.next = None # Detaches the popped node from the stack.
        self._size -= 1 # Decrements the size counter.
        return former_top.value # Returns the value of the removed node.

    def peek(self): # Method — View Top Element Without Removing
        return self._top.value if self._top else None # Inline conditional: returns value if stack is not empty, else None.

    def clear(self): # Method — Empty the Stack. This is useful for resetting the stack during reuse or testing.
        self._top = None # Resets the top pointer.
        self._size = 0 # Resets the size counter.
        return self # Enables method chaining.

    def __len__(self): # This makes your stack more Pythonic and compatible with built-in functions.
        return self._size
```

### Introduction to Queues

Queues are a fundamental data structure that follows the **First-In, First-Out (FIFO)** principle, meaning the first element added is the first one removed—like a line at a ticket counter.

Queue can again be a regular array, A singly linked list, A doubly linked list or any other linear data structure. But unlike Stacks, where one end serves as both the entry and the exit point in cues, **both ends are used**. One important distinction here is that each end serves only one purpose, either for adding elements or for removing them, **but not for both**.


**Types of Queues**
1. Simple Queue (FIFO) – Items are added at one end (rear) and removed from the other end (front).
2. Circular Queue – The last position is connected back to the first, making efficient use of memory.
3. Priority Queue – Elements are dequeued based on priority rather than order of arrival.
4. Deque (Double-Ended Queue) – Items can be added or removed from both front and rear.
	
**Operations in a Queue**
  - Enqueue: Adds an item to the rear.
  - Dequeue: Removes an item from the front.
  - Peek: Views the front item without removing it.
  - isEmpty: Checks if the queue is empty.
  - isFull (for bounded queues): Checks if the queue is full.


So let's see what choices we have for creating a queue.One possible implementation could be the regular array.We just need to enqueue at one end and dequeue from the other.But remember when we built a stack using an array, we had to think about which side to use because of performance.Keeping that in mind, let's see what options we have.

The first would be to add elements at the end of the array.This operation we already know is an O(1) complexity, so we're happy with that.When we want to remove them, the rules of queues say that we need to remove from the opposite end.So from the beginning.But we know that removing from the beginning of an array is bad for performance.So we need a better solution.And we think to ourselves, why not switch it up?Add them at the start of the array and then remove them from the end.But it's immediately obvious that this option is equally inefficient.We have an O(n) performance for adding elements at the start of an array.So regardless of which of the two solutions we choose, one operation is always going to be O(n) complexity.

In order for us to get a queue with constant performance for both Enqueue and Dequeue, we need to again turn our attention to the other data structures we learned.The linked list.A doubly linked list has O(1) performance on all methods of adding and removing from both ends,so it would naturally be a great option for a queue.A singly linked list has only one performance on three out of the four methods.The only one with O(n) is removing from the tail.

So for us to use a singly linked list as an efficient queue, we need to make sure we use the tail for adding new elements, since we know that's an O(1) complexity and the head for removing them, which is also constant time.So out of the two options we can choose from, we will go with the singly linked list

### Queues - Linked Lists Implementation

Queues are widely used in process scheduling, network management, breadth-first search (BFS), and handling requests in web servers.

For below code, we recycled it from our earlier Singly Linked List. For Enqueue we have used **SLL Append** and for Dequeue we have used **SLL Pop_left**, we have also added **tail**, replaced _length with **size**. rest everything else remains same from Stacks Linked List Implementation

```python
class Node: # Building Block of the Stack. This class enables the linked structure — each node links to the next, forming the stack.
    def __init__(self, value): # Constructor to initialize the node with a value.
        self.value = value
        self.next = None

class Queue: # Class — Linked List-Based Stack
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0  # Tracks the number of elements in the stack.
        

    def enqueue(self, value): # same as SLL Append function
        new_node = Node(value)
        if not self._size:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self._size += 1
        return self

    # This method removes the first node(head) of the list and returns its value. Its just like SLL Pop_Left
    def dequeue(self): # No  parameter is needed because we’re not adding anything, only removing.
        if not self._size:
            raise Exception("This is my custom Exception message : Queue is Empty") # If the list is empty, we cannot pop anything.Raises an exception to signal the error.
        former_head = self.head # Stores the current head node in a temp variable, so we can return its value later.This is the node we’re about to remove
        self.head = former_head.next # Moves the head pointer to the next node.Effectively removes the old head from the chain.
    # If there was only one node, former_head.next is None, so head becomes None
        former_head.next = None # Breaks the link from the removed node to the rest of the list.
        self._size -= 1
        if not self._size: # Checks again if the list is now empty (after removal).This happens if we just removed the only node
            self.tail = None #If the list is empty, set tail to None as well. Ensures both head and tail are None when the list has no nodes.
        return former_head.value # Returns the value stored in the removed node.This is the “pop” result, just like list.pop(0) in Python’s built-in list.

    def peek(self): # Method — View Top Element Without Removing
        return self.head.value if self.head else None # Inline conditional: returns value if stack is not empty, else None.

    def clear(self): # Method — Empty the Stack. This is useful for resetting the stack during reuse or testing.
        self.head = None # Resets the top pointer.
        self.tail = None
        self._size = 0 # Resets the size counter.
        return self # Enables method chaining.

    def __len__(self): # This makes your stack more Pythonic and compatible with built-in functions.
        return self._size
```

Just like stacks, queues are extensively used in the field of computer science, so much so that they have their own mathematical theory called `queueing theory`. It studies all sorts of queues and scheduling types from the simplest first in, first out, one that we just learned to more complex ones like priority queuing

**Queueing Theory**

• `Priority Queues(preemptive , non-preemptive)`
Here each member of the queue has a priority which dictates who gets served first. Priority queues can also be divided into two categories pre-emptive and non-preemptive, where the current job can or cannot be interrupted by a new element with a higher priority. A classic example would be a hospital's E.R., where patients are being treated in the order of their arrival, as long as they have the same priority. If we get, for instance, a patient with high fever and dizziness and the rest only have a mild headache or the flu, we could treat the newest arrival first, even though the other ones were in the queue for longer. If an unconscious victim of an accident were to be brought in, this new patient would have the highest priority and would even interrupt the current work of the doctors and get the required attention. So that would be an example of a pre-emptive priority queue

• `Scheduling jobs`
Other types of cues include a shortest job first Queue , where the item that would be processed the fastest is prioritized above the others, or a shortest remaining processing time cue, where the job that requires the least time to finish is served before the others

• `Messaging queues`
In today's day and age, where we do a lot of cloud computing Queues are used as means of communication between different services. They're typically called messaging queues, where all sorts of messages like requests, replies, errors or other data are being stored in a queue and will be processed in a non-blocking manner. If a server falls behind due to a sudden increase in traffic, the requests could be stored in a queue for real time delivery. Essentially saying do this work as soon as possible and no information will be lost during that moment of high load.

• `Bounded or unbounded`
Just like stacks, Bounded Queues have a preset limit of items they can hold at any time. Such a limit is normally enforced when we have limited memory. Unbounded queues, on the other hand, have no limits and can grow to as many items as the memory allows

### Queues - Deques
(the name is pronounced `deck` and is short for `double-ended queue`)

If you ever need a stack, a queue, or just a general structure where you need fast insertions and deletions from both ends, you can use a Deques. You usually don't want to build your own data structure from scratch. We did it for the purpose of learning, but in the real world you want to use something that other people have already built, especially if those people have decades of experience, like the ones who created and maintained the Python language. It will be thoroughly tested and surely more efficient than what we could build on our own

A **Deque (Double-Ended Queue)** is a special type of queue where elements can be inserted and removed from both the front and the rear. It provides more flexibility compared to standard queues.

**Types of Deques**
  1. `Input-Restricted Deque` – Insertion is allowed only at one end, but deletion can happen at both ends.
  2. `Output-Restricted Deque` – Deletion is allowed at one end, but insertion can happen at both ends.


**Operations in Deques**
* `Push Front` – Insert element at the front.
* `Push Rear` – Insert element at the rear.
* `Pop Front` – Remove element from the front.
* `Pop Rear` – Remove element from the rear.
* `Peek Front/Rear` – View the front or rear element.


**Python Implementation Using collections.deque**

Python provides an efficient built-in deque implementation.


```python
from collections import deque

# Initialize a deque
dq = deque()

# Insert elements from both ends
dq.append(10) # Push rear
dq.append(20)

dq.appendleft(5) # Push front
```
