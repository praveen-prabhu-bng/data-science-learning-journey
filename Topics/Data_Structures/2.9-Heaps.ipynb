### Introduction

So up until now, the only tree variation we talked about was the binary search tree. Its distinctive feature was that all children to the left of a node were smaller than it, and the ones to the right were larger. 

A **Heap** is also part of the tree family, which means that all the elements of a heap are represented hierarchically with parents and children. Moreover, it's also a binary tree. Therefore, any given node can have at most two children. But the similarities between the two end there because a heap is governed by different rules

**In a heap, a parent node has to be greater than all of its children and descendants.** 
For example, all the children of 27 have to be smaller than it, as well as all the nodes that branch out of 46.And this rule is applied to all the nodes of a heap. 

**The second rule is that the values of the children nodes don't dictate to which side of the parent they go on**, like they did in the binary search tree in a heap. **The larger child can be placed to either the right of its parent or to its left**. Okay, so if the value of each node is larger than the value of any of its children, this means that the root of the heap will always contain the largest value. And this is an important aspect

![](DS_Images/Heap1.png)

Let's notice another thing. Even though a parent node is always greater than its children, some nodes located higher up on a branch can be smaller than the ones at a lower level on a different branch. 

Yes, **the root will always be the largest value**, but after that, we can see in our example that at the second level of the heap we have 27, which is smaller than elements like 35,40,29. 

**So a heap does not arrange all its elements in a perfect descending order from top to bottom**. It just brings the largest value at the root level, and then it enforces the rule that every node has to be larger than its children, and by random chance, some nodes from a certain branch will end up being smaller than others from a lower branch. So this tree right here is a valid heap

**Another property of a heap is that it's perfectly balanced**. Remember that a binary search tree, or any tree in general, can be skewed to one side. In a heap, though that will never be the case because **with a heap we always complete the bottom level before starting a new one**.


![](DS_Images/Heap1.png)

So in our example, if we were to add a new element, it would go as the right child of 44, and only after the bottom level is full of nodes do we start a new one. This will make more sense when we learn how to construct a heap, but for now, just remember that a heap is a perfectly balanced tree.

So everything we learned until now in this section is true for a **max heap** where the root is the largest element. But there is also the concept of a **min heap** where the root is the smallest one, and in this case each node will be smaller in value than its children (below).

We store all the values of the heap in an array in the order from top to bottom, from left to right. So `49` is the first element of the array. Then we have `27` and `46` in this order. And then we go to the third level and get them from left to right. `10`, `22` , `42` and so on

But the next question now is how do we represent the hierarchy between the values? In a class implementation, we know that `49` has `27` and `46` as its children, because we would have the left and right properties point to those values. But how are we going to represent that with an array? We could say that `49` is the first element in the array. So it should be the root. So it should be the left child. So it has to be the right one. But what about the next ones. So it seems that we need to find a way to represent the parent child relation within the array.

To do that, we need to first remember that arrays have indices. So each item has an index representing its position in the array. And then to represent any hierarchy between these values we need a bit of help from a couple of mathematical formulas. There are two formulas which use the **index of a parent** and give us the **indices of both its children**, and they go like this

**Formula 1: Will get you the indices of the children nodes if you know the index of the parent**
![](DS_Images/Heap4.png)

**left_child_idx = 2 * parent_idx + 1**

**right_child_idx = 2 * parent_idx + 2**

So let's take this example to showcase the formulas. If we know that `27` is at index 1, the left child will be at index `1 x 2 + 1`, which is `3`. And if we look at the array we can see that `10`, which is the left child of `27` is indeed at index `3` its right child, `22` should be at index `2 x 1 + 2` which is `4`, and indeed `22` is at index `4`.

**Formula 2: Will get you the  index of the Parent nodes if you know the indices of the children nodes**
![](DS_Images/Heap5.png)

**parent_idx = (child_idx - 1) // 2**

We can find the parent by taking the index of its child, `subtract 1` from it, and then `divide the result by 2`. 

Let's take `42` as an example. If we apply the left formulas, we can find that the left child is positioned at index `11` and the right one at index `12`.But let's say we only knew the indices of these two children and want to find out the index of their parent. If we take `35`, which is at index `11`, `we subtract 1` from it, which leaves us with `10`.And then divide that by `2` we get `5`. And at index `5` we find `42`, which is indeed the parent of `35`. If we only knew the right child, which is `40`, we would take its index, which is `12`, `subtract 1` and we get `11`.Divide that by `2`.We get `5.5`. But notice that the division is not a regular one. It's a flawed division which removes any decimal values. And we're left again with `5` which is the index of `40` parent

So as you can see, even if we didn't represent the nodes as classes, we could still show the hierarchy between the elements just by storing them in an orderly fashion and applying some basic formulas.

### Insert

To summarize the process again.
1. We add the new element at the end of the array, and then we compare it with its parent.
2. If it's smaller, we swap the two values and then keep comparing until we either reach the root or until the parent is larger.
3. If the new item is smaller than its parent from the start, that's great.We don't have to do anything.


 The location of the nodes inside the heap doesn't matter. The end result we need to be concerned with is to have the largest element at the top. Heaps are generally used only to keep track of either the largest or the smallest item in a collection and then to be able to efficiently generate the next largest or smallest one if the current top is removed. So that's the theory behind inserting new elements to a heap.

We won't use node classes to store the values individually like we did in previous sections, but we do want to use a class to store the array and to keep everything tidy.

So we will have a class that will be initialized with an empty array, and then define all the functions as methods of the class.

**Insert in a Max Heap — Step-by-Step Summary**

	1. Append the Value
		○ Add the new element to the end of the array (bottom-most, right-most position in the heap).
	2. Bubble Up (move_up)
		○ Compare the new node with its parent using index math: parent_idx = (child_idx - 1) // 2
		○ If the child is greater than its parent, swap them.
		○ Update child_idx = parent_idx and repeat until the Max Heap property is restored (i.e., parent ≥ child) or the node becomes the root.
	3. Why It Works
		○ This keeps the tree complete (all levels filled left to right) and ensures that every parent is larger than its children—the defining trait of a Max Heap.


**Key Notes:**

    • The heap is implemented using a list (self.items), not a tree.
	• Index math keeps it fast and efficient: insertions take O(log n) time.
	• No need to pass the value into move_up()—it's already in the list and can be accessed via index.


Let's take an example , here we add the new element at the end of the array, and then we compare it with its parent. If it's smaller, we swap the two values and then keep comparing until we either reach the root or until the parent is larger. If the new item is smaller than its parent from the start, that's great. We don't have to do anything.

```python
class MaxBinaryHeap: #Defining a max binary heap class,which will be initialized with an items attribute.This is where all the elements will be stored
    def __init__(self):
        self.items = [] # initializing self.items as an empty list so the heap has a place to store its elements.
    # without this line, when you try to call self.items.append(...) in insert() below, Python will throw an error:
    # AttributeError: 'YourClassName' object has no attribute 'items'
    
    def insert(self, value): # Input Value = 22,since its a normal python list,We can call the append function on items and add the value as a parameter
        self.items.append(value)#first value added to the end of the list(last index).Nodes(children)created from left to right,assuming top level isfull
        self.move_up() # Then calls move_up() to restore the Max Heap property by "bubbling up" the new value if necessary.
        return self 
    # self.items is an instance variable and every method in our class can access it directly
    '''This helper function cleanly swaps two values in the list using Python’s tuple unpacking.'''
    def swap(self,idx_1,idx_2): 
        self.items[idx_1], self.items[idx_2] = self.items[idx_2], self.items[idx_1] # node 1 is assigned to node 2(using Index values) and vice versa
    '''
    Compares the newly inserted value with its parent.
    - If the child is greater, they are swapped (because it's a max heap & Parent value should be higher than child).
    - The loop continues until the new element is in the correct spot, or it becomes the root.
    '''
    def move_up(self): #Inside move_up(),the newly inserted value(22) is already sitting at the end of self.items, so there's no need to pass it in.
        child_idx = len(self.items)-1 #len(no of nodes in heap)-1,Child index set to last value(e.g 8th value assigned index 7,since index starts from 0)
        while child_idx > 0: # we have assigned an index to child and now we will identify its parent to start with the comparison.
            parent_idx = (child_idx -1) // 2 # Parent index identified as 3(10) using the formula (left child of node 10)
            if self.items[child_idx]<= self.items[parent_idx]: # Campare if the child index is less than or equal to Parent index
                break # if yes then break out of the while loop.Control returns to insert class
            self.swap(child_idx,parent_idx)
            child_idx = parent_idx  # This is the line that updates child_idx only & only IF SWAP IS REQUIRED. Now that the values have been swapped
            # the child is no more in the original index & we need to update it with current location so we can identify its new parent & compare
```

Because child_idx > 0 is still True and now parent_idx = (2 - 1) // 2 = 0

**Why didn't this happen earlier with other values?**

Because earlier inserts (like 24, 27, or 19):
   - Either didn’t require a swap
   - Or required just one swap and didn't continue bubbling upward.

With 40, it triggered two swaps—so the loop continued, and child_idx kept getting updated. That’s what changed: the inserted value was big enough to leap multiple levels.

**Why didn’t we stop earlier?**

Because the job of move_up() is to **guarantee that the inserted element reaches its correct place**, based on the Max Heap rule. And since 40 was larger than multiple ancestors (first 24, then 35), it had to bubble all the way to the root.
The loop isn't “done” until the new value **is no longer greater than its parent or it becomes the root**.

**Why This Works**

At each step:
   - The new value goes to the next available position (append)
   - It "bubbles up" if it's greater than its parent, maintaining the **Max Heap** rule.

This also shows why your move_up() doesn’t need to know the value—it just tracks where the newest node landed and compares values stored at calculated indices.


#### Time And Space Complexity

The insert method only appends the new value to the end of the list, and that is O(1) complexity. And then the move up method does some more operations. The one we need to focus on is the **while loop**, as we already know.
Loops have the potential to increase the overall time complexity by quite a lot, 

So inside the loop we do only constant time operations, some variable reassignments, and a conditional check. But the question then arises how many times does the loop run? 
The condition in our loop is the child index to reach zero, and that is even the worst case scenario, since we have a break statement inside the loop, which could stop it prematurely. 

But let's consider the worst case. So we start with child index as the last index of the list, and our loop goes at most until it reaches zero. But we don't decrement it by one. We update the child index by setting it equal to the parent index. So we essentially move in the direction down, up. We don't visit every item of the tree. Because we go from child to parent and then to parent again and so on. We only do a number of loops that is equal to the height of the tree. And if you remember from the binary search tree, when we explained the difference between a balanced tree and a skewed one, you can probably guess what the time complexity of insert is.

A heap is always perfectly balanced, so this operation of adding new elements to it will always be O(logn) complexity.

Space complexity: With each invocation of insert, we add one item to memory, and then in the helper function we create two variables and then do some reassignments and swaps which don't add anything extra to memory.
So regardless of the size of the heap, the space usage with each invocation will be constant. Therefore O(1)

##### Time :  *O* (logn)
##### Space : *O* (1)

### Remove Max

The second heap method we will discuss is remove. But unlike the method with the same name from the binary search tree, this one will not be about removing any possible element from the heap. It's only about **removing the root**. The top of the heap. This is because heaps are mostly used for placing the largest or smallest element at the top, and then having an efficient way of removing it.

So how do we remove `49`.The fastest technique that comes to mind is to just pop the first element from the list, and then somehow bring the second largest element to the top. But this method has many shortcomings. First of all, removing from the beginning of an array will require re indexing all the other elements, which is never desired. And then we have the task of rearranging the other items in a heap like manner. The solution is again, a bit counterintuitive, but actually quite a smart one once you understand it


![](DS_Images/Heap11.png)

We swap the first (`49`) and last elements(`15`) in place, and then we just remove the last one(`49`). After that, we have to restore the heap property. We first have to make sure we didn't reach the end of the array by comparing the current index with the last one. And then we need to compare the two children of the current element and swap with the largest one(`46`). And this entire logic is done in a loop until we reach the correct spot.

**Before** : Swap the top element (`42`) with last one (`19`) and then Pop the updated last element (`42`) from items list

**After**: 

![](DS_Images/Heap13.png)

```python
def swap(self,idx_1,idx_2):
    self.items[idx_1], self.items[idx_2] = self.items[idx_2], self.items[idx_1]

def remove_max(self):
    if not self.items: # checks the length of items array and if its not empty, next code(exception) will be skipped
        raise Exception("Heap is empty")
    max_elem = self.items[0] # we want to save the top element (Index 0 = Root), so we can return it in the end.
    end_idx = len(self.items)-1 # saving the end index , which is more from a readibility purposes than anything else
    self.swap(0,end_idx) # passing Index 0 and Last Index in Heap as parameters
# after the swap, next step is to remove the last element from the array, since thats where the max element is at the moment since the items array 
# is just a python list, we will use the pop() method to remove the last element (from right)
    self.items.pop()
    self.move_down()
    return max_elem #End the function by returning the root value, which should be the highest value in Max Heap

# we are now left with the task of restoring heap structure, with highest value taking the root and lower values as child     
def move_down(self):
    parent_idx = 0 # Parent index by default is assigned root index   
    child_idx = 2 * parent_idx + 1# using the formula to find the left child index using parent index,later down we will also check for right child value
    end_idx = len(self.items)-1 
    while child_idx <= end_idx:# Loop continues till such time that child index we want to swap to ends up being higher than end index
# check if child isnt the last index (< end_idx) & value not smaller than next node (right child)
# since we start with left child,check against right child ,if higher, use right child index and then go down that path of restructuring heap branch
        if child_idx < end_idx and self.items[child_idx] < self.items[child_idx + 1]:
            child_idx += 1 # use right child index if above condition is true,we assumed left child is higher, if not we will use right child index
        if self.items[parent_idx] < self.items[child.idx]: # Ensuring Parent always has higher value in Max Heap Struture
            self.swap(parent_idx, child_idx)
            parent_idx = child_idx # After values are swapped, Parent index now points to that of Child and child index updated basis new parent index
            child_idx = 2 * parent_idx + 1 
        else: 
            break # breaks out of the loop if there are no more swaps needed
```

#### Time And Space Complexity

The first steps are all constant time operations.The variable definitions, the swap, and even the pop from the list are all O(1) time complexities.So this leaves us with move down.It will define the overall time complexity of our remove max method.The first lines inside move down are also variable assignments, so nothing to see there.But then we have a loop.


We have the condition child index to be less than the end index.This means that we're increasing the child index with each iteration, and the rate at which we increase this variable should dictate the time complexity. If we look at the code at updating child_idx at the end, we're basically doubling the child index.So we're doubling the rate at which we're moving toward the end index with each iteration.
Another way of looking at this is by realizing that we're moving up and down the tree, like we did when inserting. So we're doing a number of operations that's equal to the depth of the tree. And since the heap is a perfectly balanced tree, that number is O(logn).And that is also the final complexity for the remove max.


The space complexity is again, easy to guess.We're placing a couple of variables in memory and that's about it.That number doesn't grow proportionally to the size of the heap.The space complexity is therefore O(1).

##### Time :  *O* (logn)
##### Space : *O* (1)

### Heapify using Floyd's Method


So our insert method adds new elements to an existing heap. But what if we already have a list of elements that are in random order, and we want to create a heap out of that? How would we handle such a task? I'm sure we could find a way to use the insert method we have right now to add all the elements one by one in a different list, but that would be very inefficient from both the time and memories point of view. What we really want is a function that takes an array of numbers and rearranges them in place so that they form a heap. This action is often referred to as to heapify an array, and as always, there's more than one way of achieving that.

The most famous ones would be the so-called **top down method proposed by John Williams**, and a **bottom up approach developed by Robert Floyd**, most commonly known as **Floyd's Algorithm**

so this is how the hierarchy would look like for this particular list. Now let's see how this algorithm makes a heap out of it

![](DS_Images/Heap14.png)

We start with the last node. That's not a leaf. In our case `13`. Notice that `13` and its two children actually make a mini binary tree, so nothing stops us from arranging it as a heap. And we do that with the same method we just learned in the previous section. We compare the parent node with the largest of the two children, and if it's smaller, we swap. In our case, we swap `13` with `65`.

![](DS_Images/Heap15.png)

**Below is the updated list and tree, after swapping all the individual sub trees with parent node replaced with largest child**

After we solved the entire bottom row, we moved to the next node. This method is also called a **bottom up approach**, that's because it starts with the bottom level. It makes a heap out of every individual tree and then it moves up. It solves those. And then it moves up again until the entire tree is a heap. So we have to bring it all the way down. If we notice The Root (`23`) and its two children's `44` & `11` are not in the right place. Hence we need to keep swapping till the parent is larger than its children. Hence `23` will now go way back as leaf and `65` will rise up to take the position of Root. Below is the final update Heap Tree with Index.

![](DS_Images/Heap16.png)

Let's have a look at that one more time and think about how we can implement it in code. So we said that we want to start at the last element. That's not a leaf. This means that we need a way to find its index. And there are two formulas we can use to achieve that. We can either use the index of the last element (`65`) and apply the formula to find its parent (`13`)  (formula : `parent_idx = (child_idx -1) // 2` ) or we can take the length of the array, divide that by `2` and then subtract `1`(below) . Both of them work just fine. It's just a matter of preference which one you choose

**Formula :  last_parent_idx  =  len(array) // 2 - 1**

**Before restructuring**

![](DS_Images/Heap17.png)

The formula gives us the parent for the subtree for node `65` (`formula = len(15)//2 -1 = 6`) 6th index is `13`, so we start with `13`, move the highest (`65`) as parent,
now shift left to next parent(`35`) of adjoining sub tree which is in index 5 and the subtree is restructured to replace `35` with `52`, after which we move left to next index 4 (`54`), over here the parent is correctly arranged hence we skip it and move to next index 3(`27`) and like that move to higher tree and eventually replace the root with the highest value

**After restructuring**

![](DS_Images/Heap18.png)

```python
def heapify(array): # First step is to find the index of the parent
    last_parent_idx = len(array) // 2 - 1 # calculates the index of the last parent(e.g child node 65 (index=14) parent is node 13 (index=6)
    for idx in range(last_parent_idx, -1 , -1):# start with last parent index, end at 0 Index (hence -1) and order is reverse(hence -1)
        move_down(array, idx, len(array) - 1) # idx(parent index) value will keep shifting left from one sub tree to another
    return array # return the array, which would be organised as Heap

def move_down(array, start_idx, end_idx):
    child_idx = 2 * start_idx + 1 # locating the left child of the current parent node (at start_idx).
    while child_idx <= end_idx : # The while loop uses <= to allow processing of the last possible valid child node. You enter the loop only if the 
# parent has at least one child. The if inside uses < to ensure safe access to child_idx + 1 when checking the right child.
# This loop is designed to check whether children exist

        if child_idx < end_idx and array[child_idx] < array[child_idx + 1]: # compare if right child has higher value than left child
            child_idx += 1 # if right child is higher,then we now have a final contendor to compare against parent its like eliminator round before final
            if array[start_idx] < array[child_idx]: # check if Parent node is less than child node, if yes , prepare for swap
                array[start_idx], array[child_idx] = array[child_idx] ,array[start_idx]#perform swap using tuple unpacking,right here without a fn call
            # Below two lines are the heart of how the algorithm “moves down” the tree when fixing the heap property            
            start_idx = child_idx # we bubbled up higher value to reach the root,now we need to ensure the swapped child finds the correct balance hence
# we now start tracking child index, which is now the new parent index.In other words,after a swap,the violation might have moved further down the tree. 
# So we shift our focus downward.
            child_idx = 2 * start_idx + 1 
# find new child index using updated parent index.This recalculates the left child index of the new parent (start_idx).
        else :
            break          
```

So yes — 65 is now at index 0, and 23 has moved to index 2.

Now here’s the key:
After the swap, we say:

start_idx = child_idx  # Now start_idx = 2

Because we’re not tracking the "bubbled up" value (65) — we’re tracking where 23 went, since that’s the one that might still violate heap rules further down.

**Why we reassign start_idx = child_idx**

We want to continue moving the original parent (23) down the tree. It might still be smaller than its new children, so we have to keep re-evaluating from its new position — now index 2. Think of it as tracing the path of the "offending" value until it settles into its rightful spot.


**Imagine this as a relay race:**

23 starts at root, but it's too small — so we hand the baton (23) downward to its larger child. After each swap, we follow 23 down the tree, not the one that replaced it on top.


It’s a clever optimization in the heapify algorithm! When we do:
for idx in range(last_parent_idx, -1, -1):
we're visiting parent nodes from right to left (bottom-up, reverse level order), and here's why that's smart:


**The Rationale**

In a heap (which is a complete binary tree), the bottom-rightmost nodes are the deepest and closest to being valid heaps already, because they’re either:
  1. Leaves (no children, already heap-valid), or
  2. Small subtrees that need minimal adjustment.

By processing from right to left
 - We handle subtrees from the bottom up, making sure deeper nodes are already corrected when we go higher.
 - Each move_down relies on the fact that its children are already heaps — because we’ve already visited them!

**Analogy: Building a pyramid**

Imagine you’re building a pyramid from small Lego triangles. If you start from the top, you have nowhere to place them. But if you build from the base upward — each layer solidifies the foundation for the one above it.
Same here: We heapify child subtrees first, then move up to their parents, ensuring by the time we get to the root (index 0), everything below is already perfect.one.

Imagine climbing down stairs:

  - Each swap = step down one stair.
  - At each step, you look for the next stair (child_idx).
  - If the stair number is beyond the staircase length, you stop.
So computing “stair 29” when the staircase only has 15 steps is just the way to confirm you’ve reached the bottom

#### Time And Space Complexity

Now let's talk about the performance of this algorithm.What would you guess the complexity will be?Let's see what operations we do inside Heapify.


The first thing that stands out is the for loop we do (for x in range..), where we iterate over half of the array.So that would be a total of n/2 operations, which we normally approximate to O(n).So that's the loop alone.But inside it we call "move_down" on every item we iterate over.And we know from the previous section that move_down has a complexity of log(n).

So if we have an O(logn) operation that we do in a loop, a number of O of n times, that would result in an O of n logn performance, right?

The complexity for Floyd's algorithm is actually considered to be O(n).Now let's see why. The for loop is indeed an O(n) action. But let's have a closer look at the move down method and more specifically, who we are calling it on. As we just learned, we first create a heap out of the bottom subtrees. But these are simple trees with two levels each, so the move down method will only do one swap at the most. Therefore, move down will actually run in O(1) time for all these subtrees and in a heap with n number of nodes. There are (n+1)/4 subtrees at the bottom. So basically a quarter of the time we're calling move down. It will be a constant operation. When we move one level up, it's a similar story.We have a total of (n+1)/8 sub trees where moved down does at most two swaps. Only as we get closer to the top it will truly be an O(logn) function. Coupled that with the fact that we are iterating over only half of the array, we get a final complexity for the entire algorithm bounded by O(n).

As for the space complexity, we can quickly notice by looking at the code that we're swapping items in place and not creating a new array.So the memory utilization is a constant O(1)

##### Time :  *O* (n)
##### Space : *O* (1)

### Practical Use Cases of Heapify

Heapify is the fundamental operation that builds or maintains a **binary heap**.  
This makes heaps extremely useful in several real-world applications.

---

#### 1. Priority Queues
- **Description:** Heaps are the backbone of efficient priority queues.
- **Applications:**
  - Task scheduling in operating systems.
  - Handling interrupts in hardware/software systems.
  - Event-driven simulations.

---

#### 2. Sorting (Heapsort)
- **Description:** Heapsort uses heapify to build a max-heap and repeatedly extract the largest element.
- **Applications:**
  - Sorting datasets where worst-case guarantees are important.
  - Embedded systems with limited memory (heapsort is in-place).

---

#### 3. Graph Algorithms
- **Description:** Heapify enables efficient extraction of minimum/maximum values.
- **Applications:**
  - **Dijkstra’s Algorithm**: shortest paths.
  - **Prim’s Algorithm**: minimum spanning trees.
  - **A\*** search: pathfinding in AI and robotics.

---

#### 4. Scheduling & Resource Allocation
- **Description:** Heaps manage jobs or resources in order of importance.
- **Applications:**
  - Job schedulers in distributed systems.
  - Cloud computing resource allocation.
  - Real-time systems with deadlines.

---

#### 5. Compression Algorithms
- **Description:** Huffman coding uses heaps to repeatedly extract the smallest frequency nodes.
- **Applications:**
  - File compression (ZIP, PNG).
  - Data transmission optimization.

---

#### 6. Streaming & Analytics
- **Description:** Heaps maintain the top‑k or bottom‑k elements efficiently.
- **Applications:**
  - Finding trending hashtags on social media.
  - Real-time analytics dashboards.
  - Maintaining leaderboards in gaming systems.

---

#### Summary
Heapify is not just about heapsort—it’s a **building block** for:
- Priority queues  
- Graph algorithms  
- Scheduling systems  
- Compression techniques  
- Real-time analytics
