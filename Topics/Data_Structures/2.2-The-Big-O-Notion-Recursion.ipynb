### The Big O Notation

We can have more than one algorithm that solves a given problem and they can differ in efficiency
So in order to choose the right Algorithm, How do we measure efficiency ? 
Do we time them ? 
Do we count the operations or the memory used? 

Computational complexity is the most widely accepted method for describing the efficiency of algorithms. It measures the amount of resources an algorithm requires to run. **"Big O"** is the most commonly used notation, the **"O"** stands for Order and represents the growth rate of an algorithm's runtime or space complexity in relation to the input size. It helps in analyzing how an algorithm performs as the input size scales.

Its written as a Capital O and then inside parenthesis we describe the order of growth. The order of growth is expressed in relation to a general input size of **n** Pronounced O of n, O of n Square, O of 1 etc.

For example:
- O(1) → Constant time (doesn’t depend on input size)
- O(n) → Linear time (grows proportionally with input size)
- O(n²) → Quadratic time (grows exponentially with input size)
- O(log n) → Logarithmic time (grows slowly as input increases)

Big O notation is crucial for comparing different algorithms and choosing the most efficient one for a given problem    

### Constant O(1)

Constant time complexity : describes an algorithm whose rate of growth does not change with input. 0(1) algorithms take the same amount of time and will use the same amount of memory/effort to run every time, regardless of the input size. It's generally reserved for simple functions , since more complex problems can't always be solved in constant time.

Operations with constant complexity: 
- simple variable assignments 
- arithmetic operations and comparisons 
- reference to a specific memory location (accessing elements in a list by index or in a dictionary by key)

### Linear Time O(n)

An algorithm runs in O of n if the number of operations it performs is directly proportional to the size of input. O(n) complexity indicates that the effort to run an algorithm grows proportionally to the amount of items being processed. The O of n complexity generally happens when we have to iterate over the input & do one or more constant time operations with each run. This iteration can either be a for loop or a while loop

The number of operations we do is dependent on the number of items we process. If we were to iterate over a fixed range, we wouldn’t have an O of n complexity, but rather O of 1. An O of n complexity is generally quite acceptable. It's not as efficient as the constant O of 1 we saw earlier, but it's also not as bad as some others we will see next.

Depending on the problem, we could sometimes improve an O of n solution with a bit of optimization but other times it will be the best we can possibly get

### Quadratic time O(n2)

For a given input size of n, an algorithm running in quadratic time would do n squared number of operations. Squared is not the same as multiplying, it doesn’t take much to reach some very large numbers when we square the n. O of n2 generally happens when we have one loop inside another, they can be for or while loops. O(n2) is generally viewed as pretty bad It works for small inputs, becoming unusable for large datasets 

Algorithms that run in quadratic time should be avoided if possible and most of the time we can do that by redesigning our solutions and avoiding nested loops

### Logarithmic O(logn)

In Mathematics, The logarithm is the inverse of the mathematical operation of exponentiation.
Exponentiation is the mathematical operation between 2 numbers , where one is called the base and one the power or the exponent. In our example below 2 is the base and 4 is the exponent. This operation indicates that 2 is multiplied by itself 4 times and the result of that is 16. The Logarithm is inverse of that. The operation is written as Log base 2 of 16 **($\log_{2}(16)$)**
The result of this operation is found by answering the question…2 to the power of which number is **16** and the answer is **$2^4$** written as 
$\log_{2}(16) = 4$

Another great way of thinking about logs, which will help us identify this complexity in code is to count how many times we could divide 16 by 2 and still get a number greater than or equal to 1. As highlighted below (16/2,  8/2,  4/2,  2/2) if we do it one more time we will go below 1 , so we need to stop & that was 4 times, which is exactly what Log 2 of 16 **($\log_{2}(16)$)** is

So as a final conclusion for the logarithmic complexity, like all the others, it's dependent on the data structure we apply the algorithm on. But whenever we have the possibility to achieve it, it would be worth doing it. Its running time is close to constant, even for very large datasets, and that's fantastic.

### Space Complexity

So far we have measured the efficiency of an algorithm by 2 criteria.
1. The time it takes to run  &
2. The memory It Utilizes
	
So far we have looked at the ones focusing on Time Factor . But let's now look at how the memory consumption would grow on some of these complexities
It's worth mentioning that all the concepts we learned so far still apply to Space Complexity

Auxiliary space complexity - It describes the space/memory an algorithm needs to run without counting the space/memory needed to store the input. 
As a general rule of thumb, whenever you create a new data structure inside a function, you can immediately think about the extra space you'll be using, and the order of growth for the space is determined using the same principles we used when we analyzed the time factor. Also, the variable assignment as part of the loop doesn't add up to the extra memory used, since that space is freed with each run of the loop. And that would be all about space complexity. 

### Asymptotic Complexity


Asymptotic Complexity : refers to the behavior of an algorithm as the input size grows infinitely large. It helps analyze the efficiency of algorithms without considering machine-specific details.
Key Asymptotic Notations
1. Big O (O) – Upper bound on time complexity (worst-case scenario).
2. Big Omega (Ω) – Lower bound on time complexity (best-case scenario).
3. Big Theta (Θ) – Tight bound (both upper and lower limits).


For example:
- O(n) → Linear complexity (e.g., searching in an unsorted list).
- O(log n) → Logarithmic complexity (e.g., binary search).
- O(n²) → Quadratic complexity (e.g., bubble sort).

Finding the complexity of an algorithm 
- ignore non—dominant terms 
- ignore constants 
- simple variable assignments, arithmetic operations and comparisons are usually constant time operations 
- simple loops over the input are running in 
- nested loops over the input are running in 
- logarithmic time happens when we disregard of the input with each loop linear time quadratic time large parts

### Recursion

Recursion in Python is a technique where a function calls itself to solve a problem. It helps break down complex problems into smaller, manageable subproblems.
Key Concepts of Recursion
- Base Case: The condition that stops recursion (prevents infinite loops).
- Recursive Case: The part where the function calls itself.

Recursion should always have Base case and recursive case that leads to the base one. which means if there is no fail safe, the query can run endlessly or give stack overflowasease

**Recursion vs Iteration**

When to use what ? 
- In most languages iteration is faster recursion requires call-stack management 
- recursion can use more memory because of all the stack frames 
- some languages optimize recursive calls 
- recursion usually has the advantage in code simplicity manual management of a stack in iteration can be error-prone
- As always , it depends

**"Big O" of Recursion Functions**

What to consider in determining the complexity of recursive functions:
- the number of stack frames
- the number of recursive calls
- the other operations done inside the function
- the space complexity (the max number of stack frames)
