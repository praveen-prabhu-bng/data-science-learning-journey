```python
import json, base64

# Read the image
with open("DS_Classification.png", "rb") as f:
    img_base64 = base64.b64encode(f.read()).decode("utf-8")

# Read your notebook
with open("Test_root.ipynb", "r", encoding="utf-8") as f:
    nb = json.load(f)

# Find markdown cells with the image reference and embed it
for cell in nb["cells"]:
    if cell["cell_type"] == "markdown":
        src = "".join(cell["source"])
        if "DS_Classification.png" in src:
            cell["attachments"] = {
                "DS_Classification.png": {
                    "image/png": img_base64
                }
            }
            cell["source"] = ["![](attachment:DS_Classification.png)"]
            print(f"✅ Patched cell: {cell['id']}")

# Save the patched notebook
with open("Test_root.ipynb", "w", encoding="utf-8") as f:
    json.dump(nb, f, indent=1)

print("Done! Now save, commit and push.")
```
