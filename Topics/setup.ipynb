```python
import json, base64, subprocess, os

os.chdir("D:\\Project_2026\\data-science-learning-journey")

# Fix the attachment in Test_root.ipynb
with open("Topics\\DS_Classification.png", "rb") as f:
    img = base64.b64encode(f.read()).decode()

with open("Topics\\Test_root.ipynb", "r") as f:
    nb = json.load(f)

for cell in nb["cells"]:
    if cell["cell_type"] == "markdown" and "DS_Classification" in "".join(cell["source"]):
        cell["attachments"] = {"DS_Classification.png": {"image/png": img}}
        cell["source"] = ["![](attachment:DS_Classification.png)"]
        print("Fixed!")

with open("Topics\\Test_root.ipynb", "w") as f:
    json.dump(nb, f, indent=1)

# Now convert to HTML
os.makedirs("docs", exist_ok=True)
result = subprocess.run(
    "jupyter nbconvert --to html --output-dir docs Topics\\Test_root.ipynb",
    shell=True, capture_output=True, text=True
)
print(result.stdout)
print(result.stderr)
```
