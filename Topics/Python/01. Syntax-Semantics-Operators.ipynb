####  Equality Operators

Following operations are present in python for equlity check operation-

|Operators|Meaning|
|-|-|
|**is**|***a is b*** returns true if variable/identifiers a and b *points* to the *same object*|
|**is not**|***a is not b*** returns true if variable/identifiers a and b *points* to the *different object*|
|**==**|***a == b*** returns true if variable/identifiers a and b has same value|
|**!=**|***a != b*** returns true if variable/identifiers a and b has different value|

```python
# If we create 2 variable assigning the same numbers, will each of the variable use same ram id/address or different?

lst_a=[1,2,3,4]
lst_b=[1,2,3,4]
```

```python
# checking the memory address for both variable . Diff memory address is created
# id Docstring: Return the identity of an object.
print(id(lst_a))
print(id(lst_b))
```

```python
# Using Is Operator
lst_a is lst_b
```

```python
# Now what happens when its assigned to another variable, will the address change ??
lst_a=[1,2,3,4]
lst_b=lst_a
```

```python
# They are both pointing to the same memory address
print(id(lst_a))
print(id(lst_b))
```

```python
lst_a is lst_b
```

```python
# Using Is Not function 
lst_a=[1,2,3,4]
lst_b=[1,2,3,4]
```

```python
lst_a is not lst_b
```

```python
# over here its checking the values/contents inside both variables and not the memory address
lst_a == lst_b
```
