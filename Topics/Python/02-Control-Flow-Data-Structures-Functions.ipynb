```python
# Demonstrating break, continue, and pass in one loop
for i in range(1, 9):   # Loop from 1 to 7
    if i == 3:
        print("i =", i, "→ pass (do nothing, just skip this line)")
        pass   # Does nothing, just a placeholder
    elif i == 5:
        print("i =", i, "→ continue (skip rest of this iteration)")
        continue   # Skips printing below, goes to next loop iteration
    elif i == 7:
        print("i =", i, "→ break (exit the loop completely)")
        break   # Ends the loop entirely
    print("Loop is running with i =", i)
```

- pass → Does nothing. The loop continues normally.
- continue → Skips the rest of the current iteration and jumps to the next one.
- break → Immediately stops the loop.

While pass and continue might seem similar, they actually serve different purposes in Python loops.

pass is a placeholder. It doesn't affect the loop's execution at all—it simply tells Python, "There's nothing to do here (for now), but don’t break the flow." It’s often used in function definitions or loops where code is yet to be implemented.
continue skips the rest of the loop's current iteration and moves straight to the next iteration. If Python encounters a continue statement inside a loop, it stops processing the current cycle and jumps to the next one.

##### List Comprehension
List comprehension in Python is a concise way to create new lists by looping, filtering, and transforming data in a single line. Instead of writing multiple lines with for loops, you can express the same logic more cleanly and efficiently

Syntax : [expression for item in iterable if condition]
- expression → what you want to put in the new list
- item → current element from the iterable
- iterable → sequence you’re looping over
- if condition (optional) → filter element

```python
### Basic List Comphrension
sqaure=[num**2 for num in range(10)]
print(sqaure)
```

```python
# Regular code will look like this
sqaure =[]
for num in range(10):
    sqaure.append(num**2)
print(sqaure)
```

```python
# Regular code for printing even numbers
even_numbers=[]
for num in range(10):
    if num%2==0:
        even_numbers.append(num)
print(even_numbers)
```

```python
# Using List comprehension
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
## Nested List Comphrension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## List Comprehension with function calls
words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)
```

###### Conclusion
List comprehensions are a powerful and concise way to create lists in Python. They are syntactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehensions will help you write cleaner and more efficient Python code.

#### Sets
Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

For Tuples we used (), for Lists we used [] and for Sets we use curly brackets {}

```python
##create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
## Mathematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

### Union
union_set=set1.union(set2)
print("Results of Set Union is :",union_set)

## Intersection
intersection_set=set1.intersection(set2)
print("Results of Set Intersection  is :",intersection_set)

# In Intersection_update the common elements are collected (just like in Intersection) but results are stored in first set (set1)
set1.intersection_update(set2)
print("Set1 contents after intersection are :",set1)
print("Set2 contents after intersection are :",set2)
```

```python
# Difference method
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference will remove the common elements between 2 sets and show only the remaining elements
print("Set1 contents :",set1.difference(set2))
print("Set2 contents :",set2.difference(set1))
```

```python
## Symmetric Difference - only the unique elements are combined from both the sets
set1.symmetric_difference(set2)
```

```python
## Sets Methods
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print(set1.issubset(set2)) # - issubset() → Returns True if all elements of the first set are contained in the second set.

print(set1.issuperset(set2)) # - issuperset() → Returns True if all elements of the second set are contained in the first set.
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

In this code, ** is used for dictionary unpacking in Python. It allows merging multiple dictionaries into one.
How it works:
- **dict1 and **dict2 unpack their key-value pairs into a new dictionary.
- If there are duplicate keys, values from the later dictionary (dict2) overwrite values from the earlier dictionary (dict1).

Notice how the value of "b" from dict2 (which is 3) replaces the "b" value from dict1 (2). This technique is a clean way to merge dictionaries without using loops or .update().


When merging dictionaries using {**dict1, **dict2}, the key-value pairs from dict1 are added first, followed by those from dict2. If a key exists in both dictionaries, the value from the last dictionary (dict2) overwrites the earlier one (dict1), regardless of whether the new value is higher or lower.

Tuple Assignment with * (Extended Unpacking)
Python allows you to use * when unpacking sequences to capture "the rest" of the values into a list.

What happens here:
- first → gets the first element of the tuple → 1
- last → gets the last element of the tuple → 6
- *middle → collects everything in between into a list → [2, 3, 4, 5]nces.

```python
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
#Another example
head, *body, tail = range(10)
print(head)   
print(body)   
print(tail)   
```

How It Works
This unpacking pattern is helpful when:

- You want to extract specific elements and store the rest.
- You need flexible assignment when handling unknown-length sequences.

```python
# what if i move the * order
numbers = (1, 2, 3, 4, 5, 6)

*head, middle, last = numbers

print(head)
print(middle)
print(last)
```

When you use * in unpacking:

It absorbs multiple values into a list.
- You can place it at different positions:
- first, *middle, last → first element, last element, everything else in between.
- *head, middle, last → everything except the last two goes into head.
- first, middle, *tail → first two elements assigned, rest go into tail.

So, putting * next to head means “grab all the leading values into a list, and leave the last two for middle and last.”

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ") #- end=" " keeps items on the same line separated by spaces.
    print()
```

But why arent they in the same line? 
By default, print() adds a newline ("\n") at the end of whatever it prints.When you specify end="", you’re telling Python not to add a newline, but instead use an empty string (or whatever you put there).

Answer : The inner loop uses end=" " → so items in the same sub_tuple appear on the same line separated by spaces.

Positional and Keyword arguments are two ways to pass arguments to a function.

- universally positional arguments are denoted with *args , to Pass multiple positional arguments, When unsure how many arguments will be passed

2. Keyword Arguments
- Arguments are passed using parameter names.

```python
# Combining both Positional and keyword arguments to test how system deferentiates

def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional argument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

What is map?

The map() function applies a given function to each item of an iterable (like a list, tuple, etc.) and returns a map object (which you can convert to a list, tuple, etc.)

Syntax:
map(function, iterable)
- function → the function to apply
- iterable → the sequence of value
.


```python
numbers = [1, 2, 3, 4, 5]

# Square each number
squares = map(lambda x: x**2, numbers) # - lambda x: x**2 is applied to each element in numbers.
# - The result is a map object, so we convert it to a list.
print(list(squares)) 
```

```python
### Map multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]


added_numbers= list(map(lambda x,y : x + y,numbers1,numbers2))
print(added_numbers)

# First we start with Lamda function by declaring the arguments x & y and then expression which x + y
# Second we use map , step 1 (lambda function above), followed by iterable numbers1 & numbers 2
```

Why the below code wont produce actual results

map(lambda x,y : x + y,numbers1,numbers2)

- map() in Python returns an iterator (a map object), not a list.
- When you do print(added_numbers), you’re just printing the object reference, something like: <map object at 0x000002379861D060>
    
How to see the values
You need to consume the iterator. The most common way is to WRAP IT IN "list()": 2 diff ways listed below

1. list(map(lambda x,y : x + y,numbers1,numbers2))

2. print(list(added_numbers))

Why Python does this
- In Python 3, map() was changed to return an iterator instead of a list (like in Python 2).
- This is more memory-efficient, because it doesn’t build the entire result list immediately.
- You can iterate over it once, or convert it to a list if you want all results at once.

Alternatives
You could also use:
- List comprehension (often more Pythonic):
added_numbers = [x + y for x, y in zip(numbers1, numbers2)]
print(added_numbers)  # [5, 7, 9]

Or explicitly loop through the map:
for val in map(lambda x, y: x + y, numbers1, numbers2):
    print(val)

So the short answer: map() gives you an iterator, not a list. That’s why you need list() (or another way to consume it) to see the actual values.

#### The filter() Function in Python
The filter() function constructs an iterator from elements of an iterable for which a function returns true. It is used to filter out items from a list (or any other iterable) based on a condition.

```python
# lets create a small fn to identify if a given number is true
def even(num):
    if num%2==0:
        return True

even(24)
```

```python
# Now lets combine this function with filter 
batch=[1,2,3,4,5,6,7,9,8,23,44,5,6,33,22,22]

even_num=list(filter(even,batch))

print(even_num)
```

```python
# Using filter, lambda and condition : >5
greater_than_5 =list(filter(lambda x : x > 5, batch))
print(greater_than_5)
```

```python
# Even numbers and greater than 5
greater_than_5 =list(filter(lambda x : x > 5 and x % 2 ==0 , batch))
print(greater_than_5)
```

```python
## Filter() to check if the age is greater than 25 in dictionaries
people=[
    {'name':'Praveen','age':32},
    {'name':'Jack','age':33},
    {'name':'John','age':25}
]

def age_greater_than_5(person):
    if person['age'] > 25:
        return True

list(filter(age_greater_than_5, people))
```

##### Conclusion
The filter() function is a powerful tool for creating iterators that filter items out of an iterable based on a function. It is commonly used for data cleaning, filtering objects, and removing unwanted elements from lists. By mastering filter(), you can write more concise and efficient code for processing and manipulating collections in Python.
